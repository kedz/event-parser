```python
sample_wiki = StringIO("""
<table class="metadata plainlinks ambox ambox-notice" role="presentation"><tr><td class="mbox-image"><div style="width:52px;"><a href="/wiki/File:Gnome_globe_current_event.svg" class="image"><img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/d/df/Gnome_globe_current_event.svg/40px-Gnome_globe_current_event.svg.png" width="40" height="40" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/d/df/Gnome_globe_current_event.svg/60px-Gnome_globe_current_event.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/d/df/Gnome_globe_current_event.svg/80px-Gnome_globe_current_event.svg.png 2x" /></a></div></td><td class="mbox-text"><span class="mbox-text-span">This article documents <b>&#x20;Hurricane Sandy</b>, a <b><a href="/wiki/Portal:Current_events" title="Portal:Current events">current  tropical cyclone</a></b>. Information regarding it may change rapidly as it progresses.  Although this article is updated frequently, it <a href="/wiki/Wikipedia:Risk_disclaimer" title="Wikipedia:Risk disclaimer">may not reflect</a> the most current information about this tropical cyclone for all areas.<span class="hide-when-compact"></span><span class="hide-when-compact"></span></span></td></tr></table>
<table class="infobox" border="0" cellspacing="0" cellpadding="1" style="border-collapse:collapse; margin:0 0 1em 1em;" align="right" width="280">

<tr>
<th colspan="3" style="font-size:12pt; background-color:#EEEEEE; border-bottom:1px solid #AAAAAA; padding:0;" align="center">
<table cellpadding="5" cellspacing="0" border="0" style="background-color:#EEEEEE; width:100%;">

<tr>
<th style="width:30px;">
</th>
<th> <b>Hurricane Sandy</b>
</th>
<th style="border-left:1px solid #AAAAAA; padding:5; width:24px;background: #ffffcc;"> <a href="/wiki/Saffir%E2%80%93Simpson_hurricane_wind_scale" title="Saffir–Simpson hurricane wind scale">1</a>
</th></tr></table>
</th></tr>
<tr>
<td colspan="3" align="center"><table width="280" border="0" cellspacing="0" cellpadding="4" style="background-color:#F9F9F9; border-collapse:collapse; font-size:94%; line-height:125%;">
<tr style="padding-left:0px; padding-right:9px; padding-top:3px; padding-bottom:3px;"><td align="center" valign="center"><a href="/wiki/File:Sandy_Oct_27_2012_1825Z.jpg" class="image" title="Sandy Oct 27 2012 1825Z.jpg"><img alt="Sandy Oct 27 2012 1825Z.jpg" src="//upload.wikimedia.org/wikipedia/commons/thumb/d/d4/Sandy_Oct_27_2012_1825Z.jpg/130px-Sandy_Oct_27_2012_1825Z.jpg" width="130" height="194" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/d/d4/Sandy_Oct_27_2012_1825Z.jpg/195px-Sandy_Oct_27_2012_1825Z.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/d/d4/Sandy_Oct_27_2012_1825Z.jpg/260px-Sandy_Oct_27_2012_1825Z.jpg 2x" /></a><br><small>Satellite image</td>
<td align="center" valign="center"><font/><a href="/wiki/File:18L_2012_5day.gif" class="image" title="18L 2012 5day.gif"><img alt="18L 2012 5day.gif" src="//upload.wikimedia.org/wikipedia/commons/thumb/9/9d/18L_2012_5day.gif/170px-18L_2012_5day.gif" width="170" height="136" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/9/9d/18L_2012_5day.gif/255px-18L_2012_5day.gif 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/9/9d/18L_2012_5day.gif/340px-18L_2012_5day.gif 2x" /></a><br><small>Forecast map</td></tr>
</table>
</td></tr>
<tr><th colspan="2" align="center" style="border-top:1px solid #AAAAAA; background: #ffffcc;"><b>Current storm status</b><br/>Category 1 hurricane&#160;(<a href="/wiki/Saffir%E2%80%93Simpson_Hurricane_Scale" title="Saffir–Simpson Hurricane Scale" class="mw-redirect">1-min mean</a>)</th></tr>
<tr style="background-color:#F0F0F0; border-top:1px solid #AAAAAA; border-bottom:1px solid #AAAAAA;">
<th valign="top" style="text-align:left; padding:0.1ex 0 0.1ex 5px;"> As of:
</th>
<td valign="top" style="padding:0.1ex 5px;" width="80%"> 5 p.m. <a href="/wiki/Eastern_Daylight_Time" title="Eastern Daylight Time" class="mw-redirect">EDT</a> (2100 <a href="/wiki/Coordinated_Universal_Time" title="Coordinated Universal Time">UTC</a>) October 28
</td></tr>
<tr style="border-bottom:1px solid #AAAAAA;">
<th valign="top" style="text-align:left; padding:0.1ex 0 0.1ex 5px;"> Location:
</th>
<td valign="top" style="padding:0.1ex 5px;" nowrap="nowrap" width="80%"> <span class="plainlinks nourlexpansion"><a class="external text" href="//tools.wmflabs.org/geohack/geohack.php?pagename=Hurricane_Sandy&amp;params=33.4_N_71.3_W_dim:1000km&amp;title=Hurricane+Sandy"><span class="geo-nondefault"><span class="geo-dms" title="Maps, aerial photos, and other data for this location"><span class="latitude">33°24′N</span> <span class="longitude">71°18′W</span></span></span><span class="geo-multi-punct">&#xfeff; / &#xfeff;</span><span class="geo-default"><span class="vcard"><span class="geo-dec" title="Maps, aerial photos, and other data for this location">33.4°N 71.3°W</span><span style="display:none">&#xfeff; / <span class="geo">33.4; -71.3</span></span><span style="display:none">&#xfeff; (<span class="fn org">Hurricane Sandy</span>)</span></span></span></a></span> ± 20 nm
<p>About 270 mi (435 km) <a href="/wiki/East-southeast" title="East-southeast" class="mw-redirect">ESE</a> of <a href="/wiki/Cape_Hatteras" title="Cape Hatteras">Cape Hatteras, North Carolina</a><br/>About 530 mi (850 km) <a href="/wiki/South-southeast" title="South-southeast" class="mw-redirect">SSE</a> of <a href="/wiki/New_York_City" title="New York City">New York City</a>
</p>
</td></tr>
<tr style="background-color:#F0F0F0; border-top:1px solid #AAAAAA; border-bottom:1px solid #AAAAAA;">
<th valign="top" style="text-align:left; padding:0.1ex 0 0.1ex 5px;"> Winds:
</th>
<td valign="top" style="padding:0.1ex 5px;" width="80%"> 65 <a href="/wiki/Knot_(speed)" title="Knot (speed)" class="mw-redirect">knots</a> (75 <a href="/wiki/Mph" title="Mph" class="mw-redirect">mph</a>; 120 <a href="/wiki/Km/h" title="Km/h" class="mw-redirect">km/h</a>) sustained (1-min mean)<br>gusting to 80 knots (90 mph; 150 km/h)
</td></tr>
<tr style="border-bottom:1px solid #AAAAAA;">
<th valign="top" style="text-align:left; padding:0.1ex 0 0.1ex 5px;"> Pressure:
</th>
<td valign="top" style="padding:0.1ex 5px;" nowrap="nowrap" width="80%"> 952 <a href="/wiki/Mbar" title="Mbar" class="mw-redirect">mbar</a> (<a href="/wiki/HPa" title="HPa" class="mw-redirect">hPa</a>; 28.11 <a href="/wiki/InHg" title="InHg" class="mw-redirect">inHg</a>)
</td></tr>
<tr style="background-color:#F0F0F0; border-top:1px solid #AAAAAA; border-bottom:1px solid #AAAAAA;">
<th valign="top" style="text-align:left; padding:0.1ex 0 0.1ex 5px;"> Movement:
</th>
<td valign="top" style="padding:0.1ex 5px;" width="80%"> <a href="/wiki/Northeast" title="Northeast" class="mw-redirect">NE</a> at 13 kt (15 mph; 24 km/h)
</td></tr>
<tr>
<th colspan="3" align="center" style="background: #ffffcc;"> See more <a href="#Current_storm_information">detailed information</a>.<br>
</th></tr></table>
<p><b>Hurricane Sandy</b> is a large, late-season <a href="/wiki/Tropical_cyclone" title="Tropical cyclone">tropical cyclone</a> that has affected <a href="/wiki/Jamaica" title="Jamaica">Jamaica</a>, <a href="/wiki/Cuba" title="Cuba">Cuba</a>, <a href="/wiki/The_Bahamas" title="The Bahamas">The Bahamas</a>, <a href="/wiki/Haiti" title="Haiti">Haiti</a> and <a href="/wiki/Florida" title="Florida">Florida</a>, and is currently threatening the <a href="/wiki/East_Coast_of_the_United_States" title="East Coast of the United States">East Coast of the United States</a> and <a href="/wiki/Eastern_Canada" title="Eastern Canada">Eastern Canada</a>. The eighteenth tropical cyclone, eighteenth named storm, and tenth hurricane of the <a href="/wiki/2012_Atlantic_hurricane_season" title="2012 Atlantic hurricane season">2012 Atlantic hurricane season</a>, Sandy developed from an elongated <a href="/wiki/Tropical_wave" title="Tropical wave">tropical wave</a> in the western <a href="/wiki/Caribbean_Sea" title="Caribbean Sea">Caribbean Sea</a> on October&#160;22. It quickly strengthened after becoming a tropical depression and was upgraded to a tropical storm six hours later. Sandy moved slowly northward toward the <a href="/wiki/Greater_Antilles" title="Greater Antilles">Greater Antilles</a> and gradually strengthened.
</p><p>On October&#160;24, Sandy was upgraded to a hurricane, shortly before making landfall in Jamaica. Upon moving further north, Sandy re-entered water and made its second landfall in Cuba during the early morning hours of October 25 as a Category 2 hurricane. During the late evening of October 25, Sandy weakened to Category 1 strength and headed north through the Bahamas in the early hours of October 26.<sup id="cite_ref-1" class="reference"><a href="#cite_note-1"><span>[</span>1<span>]</span></a></sup> At least 68 people were killed across the Caribbean, Bahamas, and the United States.<sup id="cite_ref-timesdispatch1_2-0" class="reference"><a href="#cite_note-timesdispatch1-2"><span>[</span>2<span>]</span></a></sup> Sandy briefly weakened to a tropical storm in the early morning hours of October 27, then restrengthened to a <a href="/wiki/Category_1_hurricane" title="Category 1 hurricane" class="mw-redirect">Category 1 hurricane</a> later that morning and remained that strength through the morning of October 28.
</p>
<div id="toc" class="toc"><div id="toctitle"><h2>Contents</h2></div>
<ul>
<li class="toclevel-1 tocsection-1"><a href="#Meteorological_history"><span class="tocnumber">1</span> <span class="toctext">Meteorological history</span></a>
<ul>
<li class="toclevel-2 tocsection-2"><a href="#Current_storm_information"><span class="tocnumber">1.1</span> <span class="toctext">Current storm information</span></a></li>
<li class="toclevel-2 tocsection-3"><a href="#Watches_and_warnings"><span class="tocnumber">1.2</span> <span class="toctext">Watches and warnings</span></a></li>
</ul>
</li>
<li class="toclevel-1 tocsection-4"><a href="#Preparation"><span class="tocnumber">2</span> <span class="toctext">Preparation</span></a>
<ul>
<li class="toclevel-2 tocsection-5"><a href="#Caribbean"><span class="tocnumber">2.1</span> <span class="toctext">Caribbean</span></a></li>
<li class="toclevel-2 tocsection-6"><a href="#The_Bahamas"><span class="tocnumber">2.2</span> <span class="toctext">The Bahamas</span></a></li>
<li class="toclevel-2 tocsection-7"><a href="#Bermuda"><span class="tocnumber">2.3</span> <span class="toctext">Bermuda</span></a></li>
<li class="toclevel-2 tocsection-8"><a href="#United_States"><span class="tocnumber">2.4</span> <span class="toctext">United States</span></a></li>
<li class="toclevel-2 tocsection-9"><a href="#Canada"><span class="tocnumber">2.5</span> <span class="toctext">Canada</span></a></li>
</ul>
</li>
<li class="toclevel-1 tocsection-10"><a href="#Impact"><span class="tocnumber">3</span> <span class="toctext">Impact</span></a>
<ul>
<li class="toclevel-2 tocsection-11"><a href="#Hispaniola"><span class="tocnumber">3.1</span> <span class="toctext">Hispaniola</span></a></li>
<li class="toclevel-2 tocsection-12"><a href="#Jamaica"><span class="tocnumber">3.2</span> <span class="toctext">Jamaica</span></a></li>
<li class="toclevel-2 tocsection-13"><a href="#Cuba"><span class="tocnumber">3.3</span> <span class="toctext">Cuba</span></a></li>
<li class="toclevel-2 tocsection-14"><a href="#Bahamas"><span class="tocnumber">3.4</span> <span class="toctext">Bahamas</span></a></li>
<li class="toclevel-2 tocsection-15"><a href="#United_States_2"><span class="tocnumber">3.5</span> <span class="toctext">United States</span></a>
<ul>
<li class="toclevel-3 tocsection-16"><a href="#Puerto_Rico"><span class="tocnumber">3.5.1</span> <span class="toctext">Puerto Rico</span></a></li>
<li class="toclevel-3 tocsection-17"><a href="#Florida"><span class="tocnumber">3.5.2</span> <span class="toctext">Florida</span></a></li>
</ul>
</li>
</ul>
</li>
<li class="toclevel-1 tocsection-18"><a href="#See_also"><span class="tocnumber">4</span> <span class="toctext">See also</span></a></li>
<li class="toclevel-1 tocsection-19"><a href="#References"><span class="tocnumber">5</span> <span class="toctext">References</span></a></li>
<li class="toclevel-1 tocsection-20"><a href="#External_links"><span class="tocnumber">6</span> <span class="toctext">External links</span></a></li>
</ul>
</div>

<h2><span class="mw-headline" id="Meteorological_history">Meteorological history</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=1" title="Edit section: Meteorological history">edit</a><span class="mw-editsection-bracket">]</span></span></h2>
<div class="thumb tleft"><div class="thumbinner" style="width:222px;"><a href="/wiki/File:Sandy_2012_track.png" class="image"><img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/9/92/Sandy_2012_track.png/220px-Sandy_2012_track.png" width="220" height="190" class="thumbimage" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/9/92/Sandy_2012_track.png/330px-Sandy_2012_track.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/9/92/Sandy_2012_track.png/440px-Sandy_2012_track.png 2x" /></a>  <div class="thumbcaption"><div class="magnify"><a href="/wiki/File:Sandy_2012_track.png" class="internal" title="Enlarge"><img src="//bits.wikimedia.org/static-1.23wmf3/skins/common/images/magnify-clip.png" width="15" height="11" alt="" /></a></div>Storm path</div></div></div>
<p>A <a href="/wiki/Tropical_wave" title="Tropical wave">tropical wave</a> was moving westward through the eastern <a href="/wiki/Caribbean_Sea" title="Caribbean Sea">Caribbean Sea</a> on October&#160;19.<sup id="cite_ref-3" class="reference"><a href="#cite_note-3"><span>[</span>3<span>]</span></a></sup> It had an extended <a href="/wiki/Low_pressure_area" title="Low pressure area" class="mw-redirect">low pressure area</a>, and conditions were expected to gradually become more favorable for <a href="/wiki/Tropical_cyclogenesis" title="Tropical cyclogenesis">development</a>.<sup id="cite_ref-4" class="reference"><a href="#cite_note-4"><span>[</span>4<span>]</span></a></sup> On October&#160;20, the system became better organized, and the  US <a href="/wiki/National_Hurricane_Center" title="National Hurricane Center">National Hurricane Center</a> (NHC) assessed a high potential for it to become a tropical cyclone within 48&#160;hours.<sup id="cite_ref-5" class="reference"><a href="#cite_note-5"><span>[</span>5<span>]</span></a></sup> By the next day, the associated <a href="/wiki/Convection_(meteorology)" title="Convection (meteorology)" class="mw-redirect">convection</a>, or thunderstorms, had become minimal, although <a href="/wiki/Atmospheric_pressure" title="Atmospheric pressure">barometric pressure</a> in the area remained low, which favored development.<sup id="cite_ref-6" class="reference"><a href="#cite_note-6"><span>[</span>6<span>]</span></a></sup> The thunderstorms gradually increased, while the system slowed and became nearly stationary over the western Caribbean.<sup id="cite_ref-7" class="reference"><a href="#cite_note-7"><span>[</span>7<span>]</span></a></sup><sup id="cite_ref-8" class="reference"><a href="#cite_note-8"><span>[</span>8<span>]</span></a></sup> At 1500&#160;<a href="/wiki/Coordinated_Universal_Time" title="Coordinated Universal Time">UTC</a> on October&#160;22, the NHC initiated advisories on Tropical Depression Eighteen about 320&#160;mi (515&#160;km) south of <a href="/wiki/Kingston,_Jamaica" title="Kingston, Jamaica">Kingston, Jamaica</a>. This was based on surface observations and satellite imagery, which indicated the system had developed enough organized convection to be classified.<sup id="cite_ref-disc1_9-0" class="reference"><a href="#cite_note-disc1-9"><span>[</span>9<span>]</span></a></sup>
</p><p>When the tropical depression formed, it was in an area of weak steering currents, located south of a <a href="/wiki/Ridge_(meteorology)" title="Ridge (meteorology)">ridge</a> extending eastward from the <a href="/wiki/Gulf_of_Mexico" title="Gulf of Mexico">Gulf of Mexico</a>. The system was in an area conducive for strengthening;<sup id="cite_ref-disc1_9-1" class="reference"><a href="#cite_note-disc1-9"><span>[</span>9<span>]</span></a></sup> this included low <a href="/wiki/Wind_shear" title="Wind shear">wind shear</a> and warm <a href="/wiki/Sea_surface_temperature" title="Sea surface temperature">sea surface temperatures</a>, and there was a possibility for <a href="/wiki/Rapid_deepening" title="Rapid deepening">rapid deepening</a>. Late on October&#160;22, a <a href="/wiki/Hurricane_Hunters" title="Hurricane Hunters">Hurricane Hunters</a> flight observed winds of 40&#160;mph (64&#160;km/h) in a <a href="/wiki/Rainband" title="Rainband">rainband</a>, which prompted NHC to upgrade the depression to Tropical Storm Sandy.<sup id="cite_ref-10" class="reference"><a href="#cite_note-10"><span>[</span>10<span>]</span></a></sup> <a href="/wiki/Outflow_(meteorology)" title="Outflow (meteorology)">Outflow</a> increased, while the convection organized further due to moist atmosphere. Due to the favorable conditions, NHC noted: "remaining nearly stationary over the warm waters of southwestern Caribbean Sea is never a good sign for this time of year."<sup id="cite_ref-11" class="reference"><a href="#cite_note-11"><span>[</span>11<span>]</span></a></sup> Despite the potential for significant intensification, the cloud pattern initially remained largely the same.<sup id="cite_ref-12" class="reference"><a href="#cite_note-12"><span>[</span>12<span>]</span></a></sup> Early on October&#160;24, an <a href="/wiki/Eye_(cyclone)" title="Eye (cyclone)">eye</a> began developing. By that time, Sandy was moving steadily northward, due to an approaching <a href="/wiki/Trough_(meteorology)" title="Trough (meteorology)">trough</a> to its northwest.<sup id="cite_ref-13" class="reference"><a href="#cite_note-13"><span>[</span>13<span>]</span></a></sup> At 11:00 a.m. EDT (1500&#160;UTC) on October&#160;24, the NHC upgraded Sandy to hurricane status after the Hurricane Hunters observed flight-level winds of 99&#160;mph (159&#160;km/h). At the time, Sandy was located about 65&#160;mi (105&#160;km) south of Kingston, Jamaica.<sup id="cite_ref-14" class="reference"><a href="#cite_note-14"><span>[</span>14<span>]</span></a></sup> 
</p>
<div class="thumb tright"><div class="thumbinner" style="width:222px;"><a href="/wiki/File:Hurricane_Sandy_GOES-13_Oct_24_2012_1445z.png" class="image"><img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/0/02/Hurricane_Sandy_GOES-13_Oct_24_2012_1445z.png/220px-Hurricane_Sandy_GOES-13_Oct_24_2012_1445z.png" width="220" height="124" class="thumbimage" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/0/02/Hurricane_Sandy_GOES-13_Oct_24_2012_1445z.png/330px-Hurricane_Sandy_GOES-13_Oct_24_2012_1445z.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/0/02/Hurricane_Sandy_GOES-13_Oct_24_2012_1445z.png/440px-Hurricane_Sandy_GOES-13_Oct_24_2012_1445z.png 2x" /></a>  <div class="thumbcaption"><div class="magnify"><a href="/wiki/File:Hurricane_Sandy_GOES-13_Oct_24_2012_1445z.png" class="internal" title="Enlarge"><img src="//bits.wikimedia.org/static-1.23wmf3/skins/common/images/magnify-clip.png" width="15" height="11" alt="" /></a></div>Satellite image of Hurricane Sandy near Jamaica landfall</div></div></div>
<p>At about 3:00 p.m. EDT (1900&#160;UTC) on October&#160;24, Sandy made <a href="/wiki/Landfall_(meteorology)" title="Landfall (meteorology)">landfall</a> near Kingston with winds of about 80&#160;mph (130&#160;km/h).<sup id="cite_ref-15" class="reference"><a href="#cite_note-15"><span>[</span>15<span>]</span></a></sup> Just offshore Cuba, Sandy rapidly intensified into a Category&#160;2 hurricane on the <a href="/wiki/Saffir-Simpson_Hurricane_Scale" title="Saffir-Simpson Hurricane Scale" class="mw-redirect">Saffir-Simpson scale</a> with 110&#160;mph (175&#160;km/h) winds.<sup id="cite_ref-Oct_25_0430_Update_16-0" class="reference"><a href="#cite_note-Oct_25_0430_Update-16"><span>[</span>16<span>]</span></a></sup> Shortly thereafter at 1:25 a.m. EDT (0525&#160;UTC), the hurricane struck Cuba just west of <a href="/wiki/Santiago_de_Cuba" title="Santiago de Cuba">Santiago de Cuba</a>.<sup id="cite_ref-17" class="reference"><a href="#cite_note-17"><span>[</span>17<span>]</span></a></sup> At landfall, Sandy had a well-defined eye of over 23&#160;mi (37&#160;km) in diameter, and flight-level winds reached 135&#160;mph (216&#160;km/h).<sup id="cite_ref-18" class="reference"><a href="#cite_note-18"><span>[</span>18<span>]</span></a></sup> While over land, the structure deteriorated and the eye was no longer visible.<sup id="cite_ref-19" class="reference"><a href="#cite_note-19"><span>[</span>19<span>]</span></a></sup> After Sandy exited Cuba, dry air and increasing shear restricted the outflow and caused the structure to become disorganized.<sup id="cite_ref-20" class="reference"><a href="#cite_note-20"><span>[</span>20<span>]</span></a></sup> A mid-level low over Florida turned the hurricane toward the north-northwest.<sup id="cite_ref-21" class="reference"><a href="#cite_note-21"><span>[</span>21<span>]</span></a></sup> By early on October&#160;26, most of the convection was sheared to the north of the center, and the size of the storm increased greatly.<sup id="cite_ref-22" class="reference"><a href="#cite_note-22"><span>[</span>22<span>]</span></a></sup> By the next day, the NHC remarked that Sandy was "showing characteristics of a hybrid cyclone... like an <a href="/wiki/Occluded_front" title="Occluded front">occluded frontal</a> low." However, the system maintained a warm thermal core, and despite strong 60&#160;mph (95&#160;km/h) wind shear, the hurricane continued to develop thunderstorms due to good <a href="/wiki/Troposphere#Divergence_and_Convergence" title="Troposphere">divergence</a> from a nearby trough; the same trough turned Sandy toward the northeast.<sup id="cite_ref-23" class="reference"><a href="#cite_note-23"><span>[</span>23<span>]</span></a></sup> On October&#160;27, Sandy briefly weakened to a tropical storm, after dry air became fully ingested into the mid- and upper-level circulations.<sup id="cite_ref-24" class="reference"><a href="#cite_note-24"><span>[</span>24<span>]</span></a></sup> Later that day, however, wind reports from the Hurricane Hunters indicated that Sandy re-intensified into a hurricane.<sup id="cite_ref-25" class="reference"><a href="#cite_note-25"><span>[</span>25<span>]</span></a></sup>
</p>
<h4><span class="mw-headline" id="Current_storm_information">Current storm information</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=2" title="Edit section: Current storm information">edit</a><span class="mw-editsection-bracket">]</span></span></h4>
<p>As of 5 p.m. <a href="/wiki/Eastern_Daylight_Time" title="Eastern Daylight Time" class="mw-redirect">EDT</a> (2100 <a href="/wiki/Coordinated_Universal_Time" title="Coordinated Universal Time">UTC</a>) October 28, Hurricane Sandy is located within 20 <a href="/wiki/Nautical_mile" title="Nautical mile">nautical miles</a> of <span class="plainlinks nourlexpansion"><a class="external text" href="//tools.wmflabs.org/geohack/geohack.php?pagename=Hurricane_Sandy&amp;params=32.4_N_71.3_W_dim:1500km&amp;title=Sandy"><span class="geo-nondefault"><span class="geo-dms" title="Maps, aerial photos, and other data for this location"><span class="latitude">32°24′N</span> <span class="longitude">71°18′W</span></span></span><span class="geo-multi-punct">&#xfeff; / &#xfeff;</span><span class="geo-default"><span class="vcard"><span class="geo-dec" title="Maps, aerial photos, and other data for this location">32.4°N 71.3°W</span><span style="display:none">&#xfeff; / <span class="geo">32.4; -71.3</span></span><span style="display:none">&#xfeff; (<span class="fn org">Sandy</span>)</span></span></span></a></span>, about 270&#160;mi (435&#160;km) <a href="/wiki/East-southeast" title="East-southeast" class="mw-redirect">east-southeast</a> of <a href="/wiki/Cape_Hatteras" title="Cape Hatteras">Cape Hatteras, North Carolina</a> and about 530&#160;mi (850&#160;km) <a href="/wiki/South-southeast" title="South-southeast" class="mw-redirect">south-southeast</a> of <a href="/wiki/New_York_City" title="New York City">New York City</a>. Maximum sustained winds are 65 knots (75&#160;mph, 120&#160;km/h), with stronger gusts. Minimum central pressure is 952 <a href="/wiki/Mbar" title="Mbar" class="mw-redirect">mbar</a> (<a href="/wiki/Pascal_(unit)" title="Pascal (unit)">hPa</a>; 28.11 <a href="/wiki/InHg" title="InHg" class="mw-redirect">InHg</a>), and the system is moving <a href="/wiki/Northeast" title="Northeast" class="mw-redirect">northeast</a> at 13 kt (15&#160;mph, 24&#160;km/h).
</p><p>Hurricane force winds extend up to 175 miles (280&#160;km) from the center of Sandy, and tropical storm force winds up to 520 miles (835&#160;km) from the center.
</p><p>For latest official information see:
</p>
<ul>
<li> The NHC's <a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/text/refresh/MIATCPAT3+shtml/">latest Public Advisory on Hurricane Sandy</a>
</li>
<li> The NHC's <a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/text/refresh/MIATCDAT3+shtml/">latest Forecast Discussion on Hurricane Sandy</a>
</li>
</ul>
<h4><span class="mw-headline" id="Watches_and_warnings">Watches and warnings</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=3" title="Edit section: Watches and warnings">edit</a><span class="mw-editsection-bracket">]</span></span></h4>
<table class="wikitable" border="0" cellspacing="0" cellpadding="0" style="font-size:95%; float:left;">







<tr>
<td style="background-color:#ffc140; width:12.6em; text-align:center; vertical-align: middle;" title="Tropical storm conditions (39-73 mph, 63-117 km/h winds) expected within 36 hours."> <b><a href="/wiki/Tropical_cyclone_warnings_and_watches#Tropical_Storm_Warning" title="Tropical cyclone warnings and watches">Tropical Storm Warning</a></b><br/><small><i>Tropical storm conditions expected within 36 hours.</i></small>
</td>
<td>
<ul>
<li> <b><a href="/wiki/United_Kingdom" title="United Kingdom">United Kingdom</a></b>
<ul>
<li> <a href="/wiki/Bermuda" title="Bermuda">Bermuda</a>
</li>
</ul>
</li>
<li> <b><a href="/wiki/United_States" title="United States">United States</a></b>
<ul>
<li> North of <a href="/wiki/Surf_City,_North_Carolina" title="Surf City, North Carolina">Surf City, North Carolina</a> to <a href="/wiki/Duck,_North_Carolina" title="Duck, North Carolina">Duck, North Carolina</a>
</li>
<li> <a href="/wiki/Pamlico_Sound" title="Pamlico Sound">Pamlico</a> and <a href="/wiki/Albemarle_Sound" title="Albemarle Sound">Albemarle</a> Sounds
</li>
</ul>
</li>
</ul>
</td></tr>

</table>
<div style="clear:both;"></div>
<p><i><b>Note</b>: In addition, <a href="/wiki/Gale_warning" title="Gale warning">high wind warnings</a> are in effect for the Mid-Atlantic states and much of New England, <a href="/wiki/Hurricane_force_wind_warning" title="Hurricane force wind warning">hurricane force wind warnings</a> are in effect from <a href="/wiki/Rhode_Island_Sound" title="Rhode Island Sound">Rhode Island Sound</a> to <a href="/wiki/Chincoteague_Bay" title="Chincoteague Bay">Chincoteague Bay</a> and <a href="/wiki/Storm_warning" title="Storm warning">storm warnings</a> are in effect from <a href="/wiki/Chincoteague_Bay" title="Chincoteague Bay">Chincoteague Bay</a> to <a href="/wiki/Currituck_Beach_Light" title="Currituck Beach Light">Currituck Beach Light</a>.</i>
</p><p><b>Forecasted impacts:</b>
</p>
<ul>
<li> Gale force winds are expected along portions of Mid-Atlantic coast Sunday (October 28), and are forecast to reach <a href="/wiki/Long_Island" title="Long Island">Long Island</a> and southern <a href="/wiki/New_England" title="New England">New England</a> by early Monday (October 29). Hurricane-force winds, at least in gusts, are likely over the warning areas and sections of the Mid-Atlantic region north of the warning areas by late Monday (October 29).
</li>
<li> A <a href="/wiki/Storm_surge" title="Storm surge">storm surge</a> is possible along the track, with the largest wave heights of between 5 to 10 feet (1.5&#160;to&#160;3.0&#160;m) from <a href="/wiki/Long_Island_Sound" title="Long Island Sound">Long Island Sound</a> to <a href="/wiki/Raritan_Bay" title="Raritan Bay">Raritan Bay</a>.
</li>
<li> Rainfall totals of 3 to 6 inches (76&#160;to&#160;150&#160;mm) are expected over far northeastern <a href="/wiki/North_Carolina" title="North Carolina">North Carolina</a>, with isolated maximum totals of 8 inches (200&#160;mm) possible. Amounts of 4 to 8 inches (100&#160;to&#160;200&#160;mm) are expected over the Mid-Atlantic states, including the <a href="/wiki/Delmarva" title="Delmarva" class="mw-redirect">Delmarva</a> peninsula with isolated maximum amounts of 12 inches (300&#160;mm) possible. Amounts of 1 to 3 inches (25&#160;to&#160;76&#160;mm) with isolated maximum amounts of 5 inches (130&#160;mm) are possible from the <a href="/wiki/Southern_Tier" title="Southern Tier">southern tier of New York</a> through New England.
</li>
<li> Researchers from <a href="/wiki/Johns_Hopkins_University" title="Johns Hopkins University">Johns Hopkins University</a>, using a computer model built on power outage data from previous hurricanes, conservatively forecast that 10 million customers along the <a href="/wiki/East_Coast_of_the_United_States" title="East Coast of the United States">Eastern Seaboard</a> will lose power from Sandy.<sup id="cite_ref-26" class="reference"><a href="#cite_note-26"><span>[</span>26<span>]</span></a></sup>
</li>
</ul>
<h2><span class="mw-headline" id="Preparation">Preparation</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=4" title="Edit section: Preparation">edit</a><span class="mw-editsection-bracket">]</span></span></h2>
<h3><span class="mw-headline" id="Caribbean">Caribbean</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=5" title="Edit section: Caribbean">edit</a><span class="mw-editsection-bracket">]</span></span></h3>
<p>Upon developing into a tropical cyclone on October&#160;22, the <a href="/wiki/Government_of_Jamaica" title="Government of Jamaica" class="mw-redirect">Government of Jamaica</a> issued a tropical storm watch for the entire island.<sup id="cite_ref-27" class="reference"><a href="#cite_note-27"><span>[</span>27<span>]</span></a></sup> Early on October&#160;23, the watch was replaced with a tropical storm warning, simultaneous to a hurricane watch being issued.<sup id="cite_ref-28" class="reference"><a href="#cite_note-28"><span>[</span>28<span>]</span></a></sup> Later that day at 1500&#160;UTC, the hurricane watch was upgraded to a hurricane warning, while the tropical storm warning was discontinued.<sup id="cite_ref-advis5_29-0" class="reference"><a href="#cite_note-advis5-29"><span>[</span>29<span>]</span></a></sup>
</p><p>Shortly after <a href="/wiki/Jamaica" title="Jamaica">Jamaica</a> began watches and warnings on October&#160;22, the <a href="/wiki/Government_of_Haiti" title="Government of Haiti">Government of Haiti</a> issued a tropical storm watch for Haiti.<sup id="cite_ref-30" class="reference"><a href="#cite_note-30"><span>[</span>30<span>]</span></a></sup> By late on October&#160;23, it was modified to a tropical storm warning.<sup id="cite_ref-31" class="reference"><a href="#cite_note-31"><span>[</span>31<span>]</span></a></sup> The <a href="/wiki/Government_of_Cuba" title="Government of Cuba" class="mw-redirect">Government of Cuba</a> posted a hurricane watch for the <a href="/wiki/Provinces_of_Cuba" title="Provinces of Cuba">Cuban Provinces</a> of <a href="/wiki/Camag%C3%BCey_Province" title="Camagüey Province">Camagüey</a>, <a href="/wiki/Granma_Province" title="Granma Province">Granma</a>, <a href="/wiki/Guant%C3%A1namo_Province" title="Guantánamo Province">Guantánamo</a>, <a href="/wiki/Holgu%C3%ADn_Province" title="Holguín Province">Holguín</a>, <a href="/wiki/Las_Tunas_Province" title="Las Tunas Province">Las Tunas</a>, and <a href="/wiki/Santiago_de_Cuba_Province" title="Santiago de Cuba Province">Santiago de Cuba</a> at 1500&#160;UTC on October&#160;23.<sup id="cite_ref-advis5_29-1" class="reference"><a href="#cite_note-advis5-29"><span>[</span>29<span>]</span></a></sup> Only three hours later, the hurricane watch was switched to a hurricane warning.<sup id="cite_ref-advis6_32-0" class="reference"><a href="#cite_note-advis6-32"><span>[</span>32<span>]</span></a></sup>
</p>
<h3><span class="mw-headline" id="The_Bahamas">The Bahamas</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=6" title="Edit section: The Bahamas">edit</a><span class="mw-editsection-bracket">]</span></span></h3>
<p>The <a href="/wiki/Government_of_the_Bahamas" title="Government of the Bahamas" class="mw-redirect">Government of the Bahamas</a>, at 1500&#160;UTC on October&#160;23, issued a tropical storm watch for several Bahamian islands, including the <a href="/wiki/Acklins" title="Acklins">Acklins</a>, <a href="/wiki/Cat_Island,_Bahamas" title="Cat Island, Bahamas">Cat Island</a>, <a href="/wiki/Crooked_Island,_Bahamas" title="Crooked Island, Bahamas">Crooked Island</a>, <a href="/wiki/Exuma" title="Exuma">Exuma</a>, <a href="/wiki/Inagua" title="Inagua">Inagua</a>, <a href="/wiki/Long_Cay" title="Long Cay">Long Cay</a>, <a href="/wiki/Long_Island,_Bahamas" title="Long Island, Bahamas">Long Island</a>, <a href="/wiki/Mayaguana" title="Mayaguana">Mayaguana</a>, <a href="/wiki/Ragged_Island,_Bahamas" title="Ragged Island, Bahamas">Ragged Island</a>, <a href="/wiki/Rum_Cay" title="Rum Cay">Rum Cay</a>, and <a href="/wiki/San_Salvador_Island" title="San Salvador Island">San Salvador Island</a>.<sup id="cite_ref-advis5_29-2" class="reference"><a href="#cite_note-advis5-29"><span>[</span>29<span>]</span></a></sup> Later that day, another tropical storm watch was issued for <a href="/wiki/Abaco_Islands" title="Abaco Islands">Abaco Islands</a>, <a href="/wiki/Andros,_Bahamas" title="Andros, Bahamas">Andros Island</a>, the <a href="/wiki/Berry_Islands" title="Berry Islands">Berry Islands</a>, <a href="/wiki/Bimini" title="Bimini">Bimini</a>, <a href="/wiki/Eleuthera" title="Eleuthera">Eleuthera</a>, <a href="/wiki/Grand_Bahama" title="Grand Bahama">Grand Bahama</a>, and <a href="/wiki/New_Providence" title="New Providence">New Providence</a>.<sup id="cite_ref-advis6_32-1" class="reference"><a href="#cite_note-advis6-32"><span>[</span>32<span>]</span></a></sup> By early on October&#160;24, the tropical storm watch for Cat Island, Exuma, Long Island, Rum Cay, and San Salvador was upgraded to a tropical storm warning.<sup id="cite_ref-33" class="reference"><a href="#cite_note-33"><span>[</span>33<span>]</span></a></sup>
</p>
<h3><span class="mw-headline" id="Bermuda">Bermuda</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=7" title="Edit section: Bermuda">edit</a><span class="mw-editsection-bracket">]</span></span></h3>
<p>At 11:15 a.m. EDT (1515&#160;UTC) on October&#160;26, the Bermuda Weather Service issued a tropical storm watch for Bermuda, reflecting the enormous size of the storm and the wide-reaching impacts.<sup id="cite_ref-34" class="reference"><a href="#cite_note-34"><span>[</span>34<span>]</span></a></sup>
</p>
<h3><span class="mw-headline" id="United_States">United States</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=8" title="Edit section: United States">edit</a><span class="mw-editsection-bracket">]</span></span></h3>
<p>Numerous schools on the <a href="/wiki/Treasure_Coast" title="Treasure Coast">Treasure Coast</a> and in <a href="/wiki/Palm_Beach_County,_Florida" title="Palm Beach County, Florida">Palm Beach County, Florida</a> announced closures for October 26, in anticipation of Sandy.<sup id="cite_ref-35" class="reference"><a href="#cite_note-35"><span>[</span>35<span>]</span></a></sup>
</p><p>Through regional offices in <a href="/wiki/Atlanta" title="Atlanta">Atlanta</a>, <a href="/wiki/Philadelphia" title="Philadelphia">Philadelphia</a>, <a href="/wiki/New_York_City" title="New York City">New York City</a> and <a href="/wiki/Boston" title="Boston">Boston</a>, <a href="/wiki/FEMA" title="FEMA" class="mw-redirect">FEMA</a> will continuously monitor Sandy and will remain in close coordination with state and tribal emergency management partners in Florida and the potentially affected Southeast, Mid-Atlantic and New England states.<sup id="cite_ref-36" class="reference"><a href="#cite_note-36"><span>[</span>36<span>]</span></a></sup>
</p><p><a href="/wiki/Amtrak" title="Amtrak">Amtrak</a> has cancelled some services through October 29 in preparation for the storm.<sup id="cite_ref-37" class="reference"><a href="#cite_note-37"><span>[</span>37<span>]</span></a></sup> 
</p><p>Flight cancellations and alerts for travel on the U.S. East Coast are in place and more may be added should the storm stay on its predicted path towards the Mid-Atlantic and the New England areas.<sup id="cite_ref-38" class="reference"><a href="#cite_note-38"><span>[</span>38<span>]</span></a></sup>
</p><p>The <a href="/wiki/National_Guard_of_the_United_States" title="National Guard of the United States">National Guard</a> and <a href="/wiki/United_States_Air_Force" title="United States Air Force">Air Force</a> put as many as 45,000 personnel in a least seven states on alert for possible duty in response to the preparations and aftermath of Sandy.<sup id="cite_ref-39" class="reference"><a href="#cite_note-39"><span>[</span>39<span>]</span></a></sup>  
</p><p>Much of the U.S. East Coast in the Mid-Atlantic and New England states have a good chance of receiving <a href="/wiki/Gale" title="Gale">gale</a>-force winds, flooding, heavy rain and possibly snow early in the week of October 28 from an unusual hybrid of Hurricane Sandy and a winter storm.<sup id="cite_ref-40" class="reference"><a href="#cite_note-40"><span>[</span>40<span>]</span></a></sup> Government weather forecasters said there is a 90% chance (the chance having increased from 60% on October 24), that the East Coast will be impacted by the storm (with some media outlets already referring to the expected combination of the tropical system and storm front under the nickname "<a href="/wiki/Frankenstein" title="Frankenstein">Frankenstorm</a>").<sup id="cite_ref-41" class="reference"><a href="#cite_note-41"><span>[</span>41<span>]</span></a></sup> Utilities and governments along the East Coast are attempting to head off long-term power failures as a result of Sandy. Power companies from the Southeast to New England are alerting <a href="/wiki/Independent_contractor" title="Independent contractor">independent contractors</a> to be ready to assist to fix storm damaged equipment quickly and are asking employees to cancel vacations and work longer hours. Baltimore Gas and Electric Co. is putting workers on standby and making plans to bring in crews from other states. In <a href="/wiki/New_Jersey" title="New Jersey">New Jersey</a>, where the storm is expected to come ashore, <a href="/wiki/Jersey_Central_Power_%26_Light" title="Jersey Central Power &amp; Light" class="mw-redirect">Jersey Central Power &amp; Light</a> has told employees to be prepared for extended shifts.<sup id="cite_ref-42" class="reference"><a href="#cite_note-42"><span>[</span>42<span>]</span></a></sup> 
</p><p>On October 26, <a href="/wiki/North_Carolina" title="North Carolina">North Carolina</a> Governor <a href="/wiki/Beverly_Perdue" title="Beverly Perdue" class="mw-redirect">Beverly Perdue</a> declared a state of emergency for 38 counties in anticipation of Hurricane Sandy, taking effect October 27.<sup id="cite_ref-43" class="reference"><a href="#cite_note-43"><span>[</span>43<span>]</span></a></sup> The <a href="/wiki/US_Navy" title="US Navy" class="mw-redirect">US Navy</a> sent more than two dozen ships to sea from <a href="/wiki/Naval_Station_Norfolk" title="Naval Station Norfolk">Norfolk Naval Base</a>, for their own protection.<sup id="cite_ref-autogenerated1_44-0" class="reference"><a href="#cite_note-autogenerated1-44"><span>[</span>44<span>]</span></a></sup>
</p><p>On October 26, <a href="/wiki/Washington,_D.C." title="Washington, D.C.">Washington, D.C.</a> Mayor <a href="/wiki/Vincent_Gray" title="Vincent Gray">Vincent Gray</a> declared a state of emergency.<sup id="cite_ref-45" class="reference"><a href="#cite_note-45"><span>[</span>45<span>]</span></a></sup> That same day the governors of <a href="/wiki/Pennsylvania" title="Pennsylvania">Pennsylvania</a>, <a href="/wiki/Maryland" title="Maryland">Maryland</a> and <a href="/wiki/Virginia" title="Virginia">Virginia</a> also declared a state of emergency in preparation of the approaching storm.<sup id="cite_ref-autogenerated1_44-1" class="reference"><a href="#cite_note-autogenerated1-44"><span>[</span>44<span>]</span></a></sup> 
</p><p>In Maryland, on October 27, <a href="/wiki/Smith_Island,_Maryland" title="Smith Island, Maryland">Smith Island</a> residents were evacuated with the assistance of the Maryland Natural Resources Police, <a href="/wiki/Dorchester_County,_Maryland" title="Dorchester County, Maryland">Dorchester County</a> opened two shelters for those in flood prone areas, and <a href="/wiki/Ocean_City,_Maryland" title="Ocean City, Maryland">Ocean City</a> initiated Phase I of their Emergency Operations Plan.<sup id="cite_ref-46" class="reference"><a href="#cite_note-46"><span>[</span>46<span>]</span></a></sup><sup id="cite_ref-47" class="reference"><a href="#cite_note-47"><span>[</span>47<span>]</span></a></sup><sup id="cite_ref-48" class="reference"><a href="#cite_note-48"><span>[</span>48<span>]</span></a></sup>  On October 28, President Obama declared an emergency in Maryland and signed an order authorized the Federal Emergency Management Agency to aid in disaster relief efforts.<sup id="cite_ref-49" class="reference"><a href="#cite_note-49"><span>[</span>49<span>]</span></a></sup>  Also, numerous areas were ordered to be evacuated including part of Ocean City, Worcester County, Wicomico County, and Somerset County.<sup id="cite_ref-50" class="reference"><a href="#cite_note-50"><span>[</span>50<span>]</span></a></sup><sup id="cite_ref-51" class="reference"><a href="#cite_note-51"><span>[</span>51<span>]</span></a></sup> 
</p><p>As of October 27, 2012, there are serious possibilities that more than a hundred million tons of dirty sediment mixed with tree limbs and debris floating behind <a href="/wiki/Conowingo_Dam" title="Conowingo Dam">Conowingo Dam</a> may be eventually poured into the <a href="/wiki/Chesapeake_Bay" title="Chesapeake Bay">Chesapeake Bay</a>, posing a potential environmental threat.<sup id="cite_ref-52" class="reference"><a href="#cite_note-52"><span>[</span>52<span>]</span></a></sup>
</p><p>In <a href="/wiki/Cape_May_County,_New_Jersey" title="Cape May County, New Jersey">Cape May County, New Jersey</a>, officials advised residents on <a href="/wiki/Barrier_islands" title="Barrier islands" class="mw-redirect">barrier islands</a> to evacuate on October 26, becoming a mandatory evacuation on October 28. There was also a voluntary evacuation for <a href="/wiki/Mantoloking,_New_Jersey" title="Mantoloking, New Jersey">Mantoloking</a>, <a href="/wiki/Bay_Head,_New_Jersey" title="Bay Head, New Jersey">Bay Head</a>, <a href="/wiki/Barnegat_Light,_New_Jersey" title="Barnegat Light, New Jersey">Barnegat Light</a>, <a href="/wiki/Beach_Haven,_New_Jersey" title="Beach Haven, New Jersey">Beach Haven</a>, <a href="/wiki/Harvey_Cedars,_New_Jersey" title="Harvey Cedars, New Jersey">Harvey Cedars</a>, <a href="/wiki/Long_Beach,_New_Jersey" title="Long Beach, New Jersey" class="mw-redirect">Long Beach</a>, <a href="/wiki/Ship_Bottom,_New_Jersey" title="Ship Bottom, New Jersey">Ship Bottom</a>, and <a href="/wiki/Stafford,_New_Jersey" title="Stafford, New Jersey" class="mw-redirect">Stafford</a> in <a href="/wiki/Ocean_County,_New_Jersey" title="Ocean County, New Jersey">Ocean County</a>.<sup id="cite_ref-Ocean_County_53-0" class="reference"><a href="#cite_note-Ocean_County-53"><span>[</span>53<span>]</span></a></sup><sup id="cite_ref-54" class="reference"><a href="#cite_note-54"><span>[</span>54<span>]</span></a></sup><sup id="cite_ref-Bay_Head_55-0" class="reference"><a href="#cite_note-Bay_Head-55"><span>[</span>55<span>]</span></a></sup> Governor <a href="/wiki/Chris_Christie" title="Chris Christie">Chris Christie</a> issued mandatory evacuations for all barrier islands from <a href="/wiki/Sandy_Hook" title="Sandy Hook">Sandy Hook</a> to <a href="/wiki/Cape_May,_New_Jersey" title="Cape May, New Jersey">Cape May</a>, along with the closure of all <a href="/wiki/Atlantic_City,_New_Jersey" title="Atlantic City, New Jersey">Atlantic City</a> casinos. Tolls were suspended on the northbound <a href="/wiki/Garden_State_Parkway" title="Garden State Parkway">Garden State Parkway</a> and the westbound <a href="/wiki/Atlantic_City_Expressway" title="Atlantic City Expressway">Atlantic City Expressway</a> starting at 6:00 a.m. on October 28.<sup id="cite_ref-56" class="reference"><a href="#cite_note-56"><span>[</span>56<span>]</span></a></sup>
</p><p><a href="/wiki/Philadelphia" title="Philadelphia">Philadelphia</a>'s Mayor <a href="/wiki/Michael_Nutter" title="Michael Nutter">Michael Nutter</a> asked residents in low-lying areas and neighborhoods prone to flooding to leave their home by 2 p.m. EDT October 28 and move to safer ground.<sup id="cite_ref-57" class="reference"><a href="#cite_note-57"><span>[</span>57<span>]</span></a></sup>
</p><p>In <a href="/wiki/New_York_City" title="New York City">New York City</a>, Mayor Michael Bloomberg told reporters on October 26 that the city has begun taking precautions but said at this time there is no call for mandatory evacuations and no plans to suspend the city's mass transit or cancel school.  On October 28 officials activated the city's coastal emergency plan, with subway closings and the evacuation of residents in areas hit during <a href="/wiki/Hurricane_Irene" title="Hurricane Irene">Hurricane Irene</a> in August 2011. Mayor <a href="/wiki/Michael_R._Bloomberg" title="Michael R. Bloomberg" class="mw-redirect">Michael R. Bloomberg</a> said that some residents should prepare to evacuate. Shelters will be opened in 65 public schools on October 28 for residents of low-lying areas who are concerned that their homes may be affected by flooding or power outages. The <a href="/wiki/Metropolitan_Transportation_Authority_(New_York)" title="Metropolitan Transportation Authority (New York)">MTA</a> announced that all subway, bus, and commuter rail service would be suspended, beginning at 7 p.m. EDT on October 28 and expected to continue through at least October 30.<sup id="cite_ref-58" class="reference"><a href="#cite_note-58"><span>[</span>58<span>]</span></a></sup> <a href="/wiki/New_York" title="New York">New York</a> Governor <a href="/wiki/Andrew_Cuomo" title="Andrew Cuomo">Andrew Cuomo</a> declared a state of emergency for every county in the state. He also asked for a pre-disaster declaration to better access federal assistance.<sup id="cite_ref-59" class="reference"><a href="#cite_note-59"><span>[</span>59<span>]</span></a></sup> On <a href="/wiki/Long_Island" title="Long Island">Long Island</a>, <a href="/wiki/Suffolk_County,_New_York" title="Suffolk County, New York">Suffolk County</a> officials ordered a mandatory evacuation of <a href="/wiki/Fire_Island" title="Fire Island">Fire Island</a> residents.
</p><p><a href="/wiki/Connecticut" title="Connecticut">Connecticut</a>'s Governor <a href="/wiki/Dan_Malloy" title="Dan Malloy" class="mw-redirect">Dan Malloy</a> partially activated the state's Emergency Operations Center on October 26,<sup id="cite_ref-60" class="reference"><a href="#cite_note-60"><span>[</span>60<span>]</span></a></sup> and signed a Declaration of Emergency the next day.<sup id="cite_ref-61" class="reference"><a href="#cite_note-61"><span>[</span>61<span>]</span></a></sup> 
</p><p>In <a href="/wiki/Rhode_Island" title="Rhode Island">Rhode Island</a>, Governor <a href="/wiki/Lincoln_Chafee" title="Lincoln Chafee">Lincoln Chafee</a> and the Rhode Island Emergency Management Agency have urged that Rhode Islanders be prepared for the storm and "pay attention to the news" as Sandy approaches.<sup id="cite_ref-chafee_62-0" class="reference"><a href="#cite_note-chafee-62"><span>[</span>62<span>]</span></a></sup><sup id="cite_ref-nationalgridri_63-0" class="reference"><a href="#cite_note-nationalgridri-63"><span>[</span>63<span>]</span></a></sup> Additionally, <a href="/wiki/National_Grid_plc#United_States" title="National Grid plc">National Grid USA</a>, Rhode Island's leading energy company, has announced that they are more well prepared than during <a href="/wiki/Hurricane_Irene" title="Hurricane Irene">Hurricane Irene</a> in 2011, when power outages were widespread in the state due to what National Grid has said was a lack of communication amongst National Grid employees.<sup id="cite_ref-nationalgridri_63-1" class="reference"><a href="#cite_note-nationalgridri-63"><span>[</span>63<span>]</span></a></sup>
</p><p>In <a href="/wiki/Massachusetts" title="Massachusetts">Massachusetts</a>, the state's emergency management agency has started issuing situational awareness news releases as some computer models show that Sandy could "potentially transition over the weekend into a powerful <a href="/wiki/Nor%27easter" title="Nor'easter">nor'easter</a>." Governor <a href="/wiki/Deval_Patrick" title="Deval Patrick">Deval Patrick</a> gave utility companies until October 26 to submit emergency plans in case the storm causes outages.<sup id="cite_ref-64" class="reference"><a href="#cite_note-64"><span>[</span>64<span>]</span></a></sup>
</p><p>On October 26, <a href="/wiki/Maine" title="Maine">Maine</a>'s Governor <a href="/wiki/Paul_LePage" title="Paul LePage">Paul LePage</a> signed a limited emergency declaration, that allows power crews from other states and Canada to help Maine prepare for Hurricane Sandy. The declaration will help Maine power providers pre-place their crews by extending the hours their crews can drive.<sup id="cite_ref-65" class="reference"><a href="#cite_note-65"><span>[</span>65<span>]</span></a></sup>
</p><p><a href="/wiki/Republican_Party_(United_States)" title="Republican Party (United States)">Republican Party</a> presidential candidate <a href="/wiki/Mitt_Romney" title="Mitt Romney">Mitt Romney</a> cancelled a campaign appearance scheduled for October 28 in <a href="/wiki/Virginia_Beach,_Virginia" title="Virginia Beach, Virginia">Virginia Beach, Virginia</a> because of Sandy, as did Vice President <a href="/wiki/Joe_Biden" title="Joe Biden">Joe Biden</a> the previous day, in order to keep law enforcement and emergency management focused on storm safety, according to their campaigns.<sup id="cite_ref-66" class="reference"><a href="#cite_note-66"><span>[</span>66<span>]</span></a></sup> President <a href="/wiki/Barack_Obama" title="Barack Obama">Barack Obama</a> canceled a campaign stop with former President <a href="/wiki/Bill_Clinton" title="Bill Clinton">Bill Clinton</a> in Virginia scheduled for October 29, as well as a trip to <a href="/wiki/Colorado_Springs,_Colorado" title="Colorado Springs, Colorado">Colorado Springs, Colorado</a>, the next day because of the impending storm, and said he will remain in Washington D.C.<sup id="cite_ref-67" class="reference"><a href="#cite_note-67"><span>[</span>67<span>]</span></a></sup>
</p><p>Over 5000 commercial airline flights scheduled for October 28 and October 29 were cancelled by the afternoon of the 28th, with that number expected to rise significantly.<sup id="cite_ref-68" class="reference"><a href="#cite_note-68"><span>[</span>68<span>]</span></a></sup>
</p><p>On October 27, <a href="/wiki/Google" title="Google">Google</a> postponed their upcoming Android event in New York City due to the storm.<sup id="cite_ref-69" class="reference"><a href="#cite_note-69"><span>[</span>69<span>]</span></a></sup>
</p><p>As of 5:35 PM EDT on October 28, the <a href="/wiki/United_States_Government" title="United States Government" class="mw-redirect">United States Government</a> <a href="/wiki/Office_of_Personnel_Management" title="Office of Personnel Management" class="mw-redirect">Office of Personnel Management</a> announced federal offices in the <a href="/wiki/Washington,_DC" title="Washington, DC" class="mw-redirect">Washington, DC</a> area were closed to the public for October 29.<sup id="cite_ref-70" class="reference"><a href="#cite_note-70"><span>[</span>70<span>]</span></a></sup> In addition, all October 29 Washington DC <a href="/wiki/Washington_Metropolitan_Area_Transit_Authority" title="Washington Metropolitan Area Transit Authority">Metro</a> service, both rail and bus, was cancelled, due to expected high winds, the likelihood of widespread power outages, and the closing of the federal government.<sup id="cite_ref-71" class="reference"><a href="#cite_note-71"><span>[</span>71<span>]</span></a></sup>
</p>
<h3><span class="mw-headline" id="Canada">Canada</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=9" title="Edit section: Canada">edit</a><span class="mw-editsection-bracket">]</span></span></h3>
<p>The <a href="/wiki/Canadian_Hurricane_Centre" title="Canadian Hurricane Centre">Canadian Hurricane Centre</a> issued its first preliminary statement on Hurricane Sandy on October 25.<sup id="cite_ref-72" class="reference"><a href="#cite_note-72"><span>[</span>72<span>]</span></a></sup> Due to the uncertainty of Sandy's path, the statement was aimed at all of Eastern Canada from the <a href="/wiki/Canadian_Maritimes" title="Canadian Maritimes" class="mw-redirect">Canadian Maritimes</a> as far west as <a href="/wiki/Southern_Ontario" title="Southern Ontario">Southern Ontario</a>. While tropical depressions occasionally reach Southern Ontario, <a href="/wiki/Hurricane_Hazel" title="Hurricane Hazel">Hurricane Hazel</a> is the only hurricane or tropical storm to have ever struck the region. Three models (out of thirty) predict landfall in <a href="/wiki/Nova_Scotia" title="Nova Scotia">Nova Scotia</a>.<sup id="cite_ref-73" class="reference"><a href="#cite_note-73"><span>[</span>73<span>]</span></a></sup> Other models, such as <a href="/wiki/The_Weather_Network" title="The Weather Network">The Weather Network</a>'s outlook, predict Sandy will slam into the Eastern Seaboard of the US, possibly affecting Southern Ontario.<sup id="cite_ref-74" class="reference"><a href="#cite_note-74"><span>[</span>74<span>]</span></a></sup>
</p>
<h2><span class="mw-headline" id="Impact">Impact</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=10" title="Edit section: Impact">edit</a><span class="mw-editsection-bracket">]</span></span></h2>
<table class="wikitable" style="float:right; margin:0 1em 0.5em;">
<caption> Effects by country
</caption>
<tr>
<th> Country </th>
<th> Fatalities </th>
<th> Damage (in USD) </th>
<th> Sources
</th></tr>
<tr>
<td> Bahamas </td>
<td style="text-align:center;"> 1  </td>
<td> Unknown </td>
<td> <sup id="cite_ref-bahd_75-0" class="reference"><a href="#cite_note-bahd-75"><span>[</span>75<span>]</span></a></sup>
</td></tr>
<tr>
<td> Cuba </td>
<td style="text-align:center;"> 11  </td>
<td> Unknown </td>
<td>  <sup id="cite_ref-Deaths_76-0" class="reference"><a href="#cite_note-Deaths-76"><span>[</span>76<span>]</span></a></sup>
</td></tr>
<tr>
<td> Dominican Republic </td>
<td style="text-align:center;"> 2 </td>
<td> Unknown </td>
<td> <sup id="cite_ref-Deaths_76-1" class="reference"><a href="#cite_note-Deaths-76"><span>[</span>76<span>]</span></a></sup>
</td></tr>
<tr>
<td> Haiti </td>
<td style="text-align:center;"> 51  </td>
<td> Unknown </td>
<td> <sup id="cite_ref-charles1_77-0" class="reference"><a href="#cite_note-charles1-77"><span>[</span>77<span>]</span></a></sup>
</td></tr>
<tr>
<td> Jamaica </td>
<td style="text-align:center;"> 1  </td>
<td> Unknown </td>
<td> <sup id="cite_ref-Deaths_76-2" class="reference"><a href="#cite_note-Deaths-76"><span>[</span>76<span>]</span></a></sup>
</td></tr>
<tr>
<td> United States </td>
<td style="text-align:center;"> 1 </td>
<td> Unknown </td>
<td> <sup id="cite_ref-timesdispatch1_2-1" class="reference"><a href="#cite_note-timesdispatch1-2"><span>[</span>2<span>]</span></a></sup>
</td></tr>
<tr>
<th> Total </th>
<th style="text-align:center;"> 67  </th>
<th> Unknown </th>
<th>
</th></tr></table>
<p>At least 68 people have been confirmed killed across the Bahamas and the Caribbean, including Cuba and Puerto Rico, as a result of the storm.<sup id="cite_ref-timesdispatch1_2-2" class="reference"><a href="#cite_note-timesdispatch1-2"><span>[</span>2<span>]</span></a></sup><sup id="cite_ref-charles1_77-1" class="reference"><a href="#cite_note-charles1-77"><span>[</span>77<span>]</span></a></sup>
</p>
<h3><span class="mw-headline" id="Hispaniola">Hispaniola</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=11" title="Edit section: Hispaniola">edit</a><span class="mw-editsection-bracket">]</span></span></h3>
<p>In Haiti, which was still recovering from the <a href="/wiki/2010_Haiti_earthquake" title="2010 Haiti earthquake">2010 earthquake</a>, at least 51 people have died and an estimated 200,000 were left homeless as of late Saturday October 27 as a result of four days of ongoing rain from Hurricane Sandy.<sup id="cite_ref-charles1_77-2" class="reference"><a href="#cite_note-charles1-77"><span>[</span>77<span>]</span></a></sup> Reports of significant damage to <a href="/wiki/Port-Salut" title="Port-Salut">Port-Salut</a> were received as rivers overflowed their banks.<sup id="cite_ref-78" class="reference"><a href="#cite_note-78"><span>[</span>78<span>]</span></a></sup> In the capital of <a href="/wiki/Port-au-Prince" title="Port-au-Prince">Port-au-Prince</a>, whole streets were flooded by the heavy rains. Most of the tents and buildings in the city's sprawling refugee camps and the <a href="/wiki/Cite_Soleil" title="Cite Soleil" class="mw-redirect">Cite Soleil</a> neighborhood were flooded or leaking, a repeat of what happened earlier in the year during the passage of <a href="/wiki/Hurricane_Isaac_(2012)" title="Hurricane Isaac (2012)">Hurricane Isaac</a>.<sup id="cite_ref-autogenerated2_79-0" class="reference"><a href="#cite_note-autogenerated2-79"><span>[</span>79<span>]</span></a></sup>
</p><p>In the Dominican Republic, two people were killed and 8,755 people evacuated as officials said the rains were expected to continue until at least Saturday.<sup id="cite_ref-Deaths_76-3" class="reference"><a href="#cite_note-Deaths-76"><span>[</span>76<span>]</span></a></sup><sup id="cite_ref-80" class="reference"><a href="#cite_note-80"><span>[</span>80<span>]</span></a></sup>  Traveling by vehicle was very hard in places as some roads had high water levels. An employee of <a href="/wiki/CNN" title="CNN">CNN</a> estimated 70% of the streets in <a href="/wiki/Santo_Domingo" title="Santo Domingo">Santo Domingo</a> were flooded. Some cars were underwater and people with trucks were charging motorists $5 to pull their vehicles out, while others were doing it for free.<sup id="cite_ref-81" class="reference"><a href="#cite_note-81"><span>[</span>81<span>]</span></a></sup>
</p>
<h3><span class="mw-headline" id="Jamaica">Jamaica</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=12" title="Edit section: Jamaica">edit</a><span class="mw-editsection-bracket">]</span></span></h3>
<p>Sandy was the first direct hit by the eye of a hurricane on Jamaica since <a href="/wiki/Hurricane_Gilbert" title="Hurricane Gilbert">Hurricane Gilbert</a> 24 years ago. The storm hit Jamaica as a category 1 hurricane. Extensive damage was reported on the island. Trees and power lines were snapped and shanty houses were heavily damaged, both from the winds and flooding rains. More than 100 fishermen were stranded in outlying <a href="/wiki/Pedro_Cays" title="Pedro Cays" class="mw-redirect">Pedro Cays</a> off Jamaica's southern coast.<sup id="cite_ref-82" class="reference"><a href="#cite_note-82"><span>[</span>82<span>]</span></a></sup> One man was crushed to death by stones that fell from a hillside as he tried to get into his house in a rural village near Kingston.<sup id="cite_ref-83" class="reference"><a href="#cite_note-83"><span>[</span>83<span>]</span></a></sup>  The country's sole electricity provider, the Jamaica Public Service Company, reported that 70 percent of its customers were without power. Looters shot and wounded a police official as he led a group of officers through Craig Town, a section of West Kingston. More than 1,000 people went to shelters, the Office of Disaster Preparedness said. Jamaican authorities closed the island's international airports and police ordered 48-hour curfews in major towns to keep people off the streets and deter looting. Cruise ships changed their itineraries to avoid the storm, which made landfall Wednesday afternoon near the capital, Kingston.<sup id="cite_ref-84" class="reference"><a href="#cite_note-84"><span>[</span>84<span>]</span></a></sup>
</p><p>The day after the storm, government officials went on an aerial tour of the rural eastern areas of the island. Parliament member Daryl Vaz reported that most buildings had lost their roofs, in addition to widespread damage to banana crops. Approximately 70 percent of the island lost power because of Sandy, and schools in the Kingston area would likely remain closed for a week. Resorts in <a href="/wiki/Montego_Bay" title="Montego Bay">Montego Bay</a> and <a href="/wiki/Negril" title="Negril">Negril</a> sustained no major damage, and cruise ship terminals reopened to vessels after a 24-hour suspension of services. Authorities warned that the extent of the damage is not clear, since some major roads remained impassable, and it would likely be weeks before life in most areas returned to normal.<sup id="cite_ref-autogenerated2_79-1" class="reference"><a href="#cite_note-autogenerated2-79"><span>[</span>79<span>]</span></a></sup>
</p>
<h3><span class="mw-headline" id="Cuba">Cuba</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=13" title="Edit section: Cuba">edit</a><span class="mw-editsection-bracket">]</span></span></h3>
<div class="thumb tright"><div class="thumbinner" style="width:222px;"><a href="/wiki/File:Flickr_-_DVIDSHUB_-_Hurricane_Sandy_hits_Guantanamo_Bay_(Image_1_of_3).jpg" class="image"><img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/f/fb/Flickr_-_DVIDSHUB_-_Hurricane_Sandy_hits_Guantanamo_Bay_%28Image_1_of_3%29.jpg/220px-Flickr_-_DVIDSHUB_-_Hurricane_Sandy_hits_Guantanamo_Bay_%28Image_1_of_3%29.jpg" width="220" height="146" class="thumbimage" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/f/fb/Flickr_-_DVIDSHUB_-_Hurricane_Sandy_hits_Guantanamo_Bay_%28Image_1_of_3%29.jpg/330px-Flickr_-_DVIDSHUB_-_Hurricane_Sandy_hits_Guantanamo_Bay_%28Image_1_of_3%29.jpg 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/f/fb/Flickr_-_DVIDSHUB_-_Hurricane_Sandy_hits_Guantanamo_Bay_%28Image_1_of_3%29.jpg/440px-Flickr_-_DVIDSHUB_-_Hurricane_Sandy_hits_Guantanamo_Bay_%28Image_1_of_3%29.jpg 2x" /></a>  <div class="thumbcaption"><div class="magnify"><a href="/wiki/File:Flickr_-_DVIDSHUB_-_Hurricane_Sandy_hits_Guantanamo_Bay_(Image_1_of_3).jpg" class="internal" title="Enlarge"><img src="//bits.wikimedia.org/static-1.23wmf3/skins/common/images/magnify-clip.png" width="15" height="11" alt="" /></a></div>Hurricane Sandy damage in Guantanamo Bay</div></div></div>
<p>Hurricane Sandy strengthened into a Category 2 hurricane before hitting Cuba.<sup id="cite_ref-Oct_25_0430_Update_16-1" class="reference"><a href="#cite_note-Oct_25_0430_Update-16"><span>[</span>16<span>]</span></a></sup> At least 55,000 people had been evacuated principally because of expected flooding from rains that could total up to 20 inches (500&#160;mm) in some places and a storm surge the Cuban weather service said was already beginning along the southeastern coast around midnight EDT.<sup id="cite_ref-85" class="reference"><a href="#cite_note-85"><span>[</span>85<span>]</span></a></sup> Sandy made landfall just west of Santiago de Cuba, the country's second-largest city, as a strong Category 2 hurricane, with the strong eastern eyewall passing directly over the city.<sup id="cite_ref-86" class="reference"><a href="#cite_note-86"><span>[</span>86<span>]</span></a></sup><sup id="cite_ref-Hurricane_Sandy_Update_Oct_25_0530_UTC_87-0" class="reference"><a href="#cite_note-Hurricane_Sandy_Update_Oct_25_0530_UTC-87"><span>[</span>87<span>]</span></a></sup>  The eye of the storm came ashore just west of the city with waves up to 29 feet (9 meters) and a six-foot (2 meter) storm surge that caused extensive coastal flooding.<sup id="cite_ref-88" class="reference"><a href="#cite_note-88"><span>[</span>88<span>]</span></a></sup> 
</p><p>Reports from the area after the passage of Sandy spoke of widespread damage, particularly to Santiago de Cuba. Electricity and water services had been knocked out, and most of the trees in the city had either been ripped off their roots or had lost all their leaves. A local TV reporter told Cuban state television that many of the city's 300,000 homes were reportedly in bad shape before the storm and were left vulnerable to high winds and heavy rain. Several Cuban provinces promised to send brigades to help Santiago recover, although officials gave a long list of other towns that suffered devastation. <a href="/wiki/Guantanamo" title="Guantanamo" class="mw-redirect">Guantanamo</a> followed a similar fate to Santiago, with television showing telephone poles and cables down across the city. Several historic buildings in the center of town were reportedly damaged. 
State media has said at least 11 people in Cuba were killed as a result of the storm, and <a href="/wiki/Raul_Castro" title="Raul Castro" class="mw-redirect">Raul Castro</a> planned to visit Santiago de Cuba in the coming days. Nine of the deaths were in <a href="/wiki/Santiago_de_Cuba_Province" title="Santiago de Cuba Province">Santiago de Cuba Province</a> and two were in <a href="/wiki/Guantanamo_Province" title="Guantanamo Province" class="mw-redirect">Guantanamo Province</a> and most of the victims were trapped in destroyed houses.<sup id="cite_ref-89" class="reference"><a href="#cite_note-89"><span>[</span>89<span>]</span></a></sup><sup id="cite_ref-90" class="reference"><a href="#cite_note-90"><span>[</span>90<span>]</span></a></sup> This makes Sandy the deadliest hurricane to hit Cuba since 2005, when <a href="/wiki/Hurricane_Dennis" title="Hurricane Dennis">Hurricane Dennis</a> killed 16 people.<sup id="cite_ref-91" class="reference"><a href="#cite_note-91"><span>[</span>91<span>]</span></a></sup> 
</p><p>Damage to the U.S. <a href="/wiki/Guantanamo_Bay_Naval_Base" title="Guantanamo Bay Naval Base">Guantanamo Bay Naval Base</a> was not as severe and there were no reports of injuries at the base. The highest sustained winds were below hurricane strength at 54 miles per hour (87&#160;km/h), with a maximum gust of 66 miles per hour (106&#160;km/h). The storm damaged roofs and windows in a few older buildings, and tore some of the power cables within the facility. Several recreational boats broke off their moorings, but there was no damage to the prison, according to Navy Capt. Robert Durand.<sup id="cite_ref-92" class="reference"><a href="#cite_note-92"><span>[</span>92<span>]</span></a></sup>
</p>
<h3><span class="mw-headline" id="Bahamas">Bahamas</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=14" title="Edit section: Bahamas">edit</a><span class="mw-editsection-bracket">]</span></span></h3>
<p>A <a href="/wiki/NOAA" title="NOAA" class="mw-redirect">NOAA</a> automated station at <a href="/wiki/West_End,_Grand_Bahama" title="West End, Grand Bahama">Settlement Point</a> on <a href="/wiki/Grand_Bahama_Island" title="Grand Bahama Island" class="mw-redirect">Grand Bahama Island</a> reported sustained winds of 49&#160;mph (74&#160;km/h) and a wind gust of 63&#160;mph (102&#160;km/h).<sup id="cite_ref-93" class="reference"><a href="#cite_note-93"><span>[</span>93<span>]</span></a></sup> One person was killed in <a href="/wiki/Lyford_Cay" title="Lyford Cay">Lyford Cay</a> after being blown off his porch.<sup id="cite_ref-bahd_75-1" class="reference"><a href="#cite_note-bahd-75"><span>[</span>75<span>]</span></a></sup> Portions of the Bahamas lost power or cellular service, including an islandwide power outage on <a href="/wiki/Bimini" title="Bimini">Bimini</a>. Five homes were severely damaged near Williams's Town.<sup id="cite_ref-94" class="reference"><a href="#cite_note-94"><span>[</span>94<span>]</span></a></sup>
</p>
<h3><span class="mw-headline" id="United_States_2">United States</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=15" title="Edit section: United States">edit</a><span class="mw-editsection-bracket">]</span></span></h3>
<h4><span class="mw-headline" id="Puerto_Rico">Puerto Rico</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=16" title="Edit section: Puerto Rico">edit</a><span class="mw-editsection-bracket">]</span></span></h4>
<p>Police said a man was killed on October 26 in <a href="/wiki/Juana_D%C3%ADaz,_Puerto_Rico" title="Juana Díaz, Puerto Rico">Juana Diaz</a>. He was swept away in a river swollen by rain from Sandy's outer bands. In addition, flooding forced at least 100 families in the southwest to seek new shelter.<sup id="cite_ref-timesdispatch1_2-3" class="reference"><a href="#cite_note-timesdispatch1-2"><span>[</span>2<span>]</span></a></sup>
</p>
<h4><span class="mw-headline" id="Florida">Florida</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=17" title="Edit section: Florida">edit</a><span class="mw-editsection-bracket">]</span></span></h4>
<p>High winds and waves washed sand onto coastal roads in southeastern Florida. The storm left power outages across the region, which left many traffic lights out of order.<sup id="cite_ref-95" class="reference"><a href="#cite_note-95"><span>[</span>95<span>]</span></a></sup> The <a href="/wiki/Coast_Guard" title="Coast Guard" class="mw-redirect">Coast Guard</a> rescued two sea men in <a href="/wiki/Volusia_County" title="Volusia County" class="mw-redirect">Volusia County</a> off <a href="/wiki/New_Smyrna_Beach" title="New Smyrna Beach" class="mw-redirect">New Smyrna Beach</a> on the morning of October 26.<sup id="cite_ref-96" class="reference"><a href="#cite_note-96"><span>[</span>96<span>]</span></a></sup> <a href="/wiki/Brevard_County" title="Brevard County" class="mw-redirect">Brevard County</a> and Volusia County Schools cancelled all extracurricular activities for October 26, including Friday night football.<sup id="cite_ref-97" class="reference"><a href="#cite_note-97"><span>[</span>97<span>]</span></a></sup> In east-central Florida, damage was minor, though the storm left about 1,000&#160;people without power.<sup id="cite_ref-Orlando_98-0" class="reference"><a href="#cite_note-Orlando-98"><span>[</span>98<span>]</span></a></sup> Airlines at <a href="/wiki/Miami_International_Airport" title="Miami International Airport">Miami International Airport</a> canceled more than 20 flights to or from <a href="/wiki/Jamaica" title="Jamaica">Jamaica</a> or <a href="/wiki/The_Bahamas" title="The Bahamas">the Bahamas</a>, while some airlines flying from <a href="/wiki/Fort_Lauderdale%E2%80%93Hollywood_International_Airport" title="Fort Lauderdale–Hollywood International Airport">Fort Lauderdale–Hollywood International Airport</a> canceled a total of 13 flights to the islands.<sup id="cite_ref-autogenerated1_44-2" class="reference"><a href="#cite_note-autogenerated1-44"><span>[</span>44<span>]</span></a></sup>
</p>
<h2><span class="mw-headline" id="See_also">See also</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=18" title="Edit section: See also">edit</a><span class="mw-editsection-bracket">]</span></span></h2>
<div class="noprint tright portal" style="border:solid #aaa 1px;margin:0.5em 0 0.5em 1em;">
<table style="background:#f9f9f9;font-size:85%;line-height:110%;max-width:175px;">
<tr valign="middle"><td style="text-align:center;"><a href="/wiki/File:Cyclone_Catarina_from_the_ISS_on_March_26_2004.JPG" class="image"><img alt="Portal icon" src="//upload.wikimedia.org/wikipedia/commons/thumb/8/89/Cyclone_Catarina_from_the_ISS_on_March_26_2004.JPG/32px-Cyclone_Catarina_from_the_ISS_on_March_26_2004.JPG" width="32" height="21" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/8/89/Cyclone_Catarina_from_the_ISS_on_March_26_2004.JPG/48px-Cyclone_Catarina_from_the_ISS_on_March_26_2004.JPG 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/8/89/Cyclone_Catarina_from_the_ISS_on_March_26_2004.JPG/64px-Cyclone_Catarina_from_the_ISS_on_March_26_2004.JPG 2x" /></a></td><td style="padding:0 0.2em;vertical-align:middle;font-style:italic;font-weight:bold;"><a href="/wiki/Portal:Tropical_cyclones" title="Portal:Tropical cyclones">Tropical cyclones portal</a></td></tr></table></div>
<ul>
<li><a href="/wiki/List_of_North_Carolina_hurricanes" title="List of North Carolina hurricanes">List of North Carolina hurricanes</a>
</li>
<li><a href="/wiki/List_of_Virginia_hurricanes" title="List of Virginia hurricanes" class="mw-redirect">List of Virginia hurricanes</a>
</li>
<li><a href="/wiki/1991_Perfect_Storm" title="1991 Perfect Storm">1991 Perfect Storm</a>, similar setup in the eastern United States
</li>
<li><a href="/wiki/December_1992_nor%27easter" title="December 1992 nor'easter">December 1992 nor'easter</a>, last severe flood event in New York City
</li>
<li><a href="/wiki/Hurricane_Hazel" title="Hurricane Hazel">Hurricane Hazel</a>, another complicated trough interaction storm in eastern North America
</li>
<li><a href="/wiki/Hurricane_Irene" title="Hurricane Irene">Hurricane Irene</a>, last hurricane to hit the <a href="/wiki/Mid-Atlantic_states" title="Mid-Atlantic states">Mid-Atlantic states</a>
</li>
<li><a href="/wiki/Hurricane_Paloma" title="Hurricane Paloma">Hurricane Paloma</a>, similar late season intense storm in Cuba
</li>
<li><a href="/wiki/List_of_New_Jersey_hurricanes" title="List of New Jersey hurricanes">List of New Jersey hurricanes</a>
</li>
<li><a href="/wiki/List_of_New_York_hurricanes" title="List of New York hurricanes">List of New York hurricanes</a>
</li>
</ul>
<h2><span class="mw-headline" id="References">References</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=19" title="Edit section: References">edit</a><span class="mw-editsection-bracket">]</span></span></h2>
<div class="reflist columns references-column-width" style="-moz-column-width: 30em; -webkit-column-width: 30em; column-width: 30em; list-style-type: decimal;">
<ol class="references">
<li id="cite_note-1"><span class="mw-cite-backlink"><b><a href="#cite_ref-1">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.cfnews13.com/content/news/cfnews13/news/article.html/content/news/articles/cfn/2012/10/25/hurricane_sandy_movi.html?cmpid=twitter">"Hurricane Sandy storms through Bahamas, Central Florida on alert"</a>. Central Florida News 13<span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Hurricane+Sandy+storms+through+Bahamas%2C+Central+Florida+on+alert&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.cfnews13.com%2Fcontent%2Fnews%2Fcfnews13%2Fnews%2Farticle.html%2Fcontent%2Fnews%2Farticles%2Fcfn%2F2012%2F10%2F25%2Fhurricane_sandy_movi.html%3Fcmpid%3Dtwitter&amp;rft.pub=Central+Florida+News+13&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-timesdispatch1-2"><span class="mw-cite-backlink">^ <a href="#cite_ref-timesdispatch1_2-0"><sup><i><b>a</b></i></sup></a> <a href="#cite_ref-timesdispatch1_2-1"><sup><i><b>b</b></i></sup></a> <a href="#cite_ref-timesdispatch1_2-2"><sup><i><b>c</b></i></sup></a> <a href="#cite_ref-timesdispatch1_2-3"><sup><i><b>d</b></i></sup></a></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www2.timesdispatch.com/news/2012/oct/28/tdmain16-states-declare-storm-emergency-ar-2316785/">"States declare storm emergency"</a>. Richmond Times Dispatch<span class="reference-accessdate">. Retrieved 28 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=States+declare+storm+emergency&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww2.timesdispatch.com%2Fnews%2F2012%2Foct%2F28%2Ftdmain16-states-declare-storm-emergency-ar-2316785%2F&amp;rft.pub=Richmond+Times+Dispatch&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-3"><span class="mw-cite-backlink"><b><a href="#cite_ref-3">^</a></b></span> <span class="reference-text"><span class="citation report">Todd Kimberlain&#32;(2012-10-19)&#32;(TXT).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210191736.txt">&#xfeff;Tropical Weather Outlook&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210191736.txt">http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210191736.txt</a></span><span class="reference-accessdate">.&#32;Retrieved 2012-10-23</span>.</span></span>
</li>
<li id="cite_note-4"><span class="mw-cite-backlink"><b><a href="#cite_ref-4">^</a></b></span> <span class="reference-text"><span class="citation report">Stacy Stewart&#32;(2012-10-19)&#32;(TXT).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210192339.txt">&#xfeff;Tropical Weather Outlook&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210192339.txt">http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210192339.txt</a></span><span class="reference-accessdate">.&#32;Retrieved 2012-10-23</span>.</span></span>
</li>
<li id="cite_note-5"><span class="mw-cite-backlink"><b><a href="#cite_ref-5">^</a></b></span> <span class="reference-text"><span class="citation report">Eric Blake&#32;(2012-10-20)&#32;(TXT).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210201752.txt">&#xfeff;Tropical Weather Outlook&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210201752.txt">http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210201752.txt</a></span><span class="reference-accessdate">.&#32;Retrieved 2012-10-23</span>.</span></span>
</li>
<li id="cite_note-6"><span class="mw-cite-backlink"><b><a href="#cite_ref-6">^</a></b></span> <span class="reference-text"><span class="citation report">Jack Beven&#32;(2012-10-21)&#32;(TXT).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210210535.txt">&#xfeff;Tropical Weather Outlook&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210210535.txt">http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210210535.txt</a></span><span class="reference-accessdate">.&#32;Retrieved 2012-10-23</span>.</span></span>
</li>
<li id="cite_note-7"><span class="mw-cite-backlink"><b><a href="#cite_ref-7">^</a></b></span> <span class="reference-text"><span class="citation report">Robbie Berg&#32;(2012-10-21)&#32;(TXT).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210211132.txt">&#xfeff;Tropical Weather Outlook&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210211132.txt">http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210211132.txt</a></span><span class="reference-accessdate">.&#32;Retrieved 2012-10-23</span>.</span></span>
</li>
<li id="cite_note-8"><span class="mw-cite-backlink"><b><a href="#cite_ref-8">^</a></b></span> <span class="reference-text"><span class="citation report">Stacy Stewart&#32;(2012-10-22)&#32;(TXT).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210220555.txt">&#xfeff;Tropical Weather Outlook&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210220555.txt">http://www.nhc.noaa.gov/archive/text/TWOAT/2012/TWOAT.201210220555.txt</a></span><span class="reference-accessdate">.&#32;Retrieved 2012-10-23</span>.</span></span>
</li>
<li id="cite_note-disc1-9"><span class="mw-cite-backlink">^ <a href="#cite_ref-disc1_9-0"><sup><i><b>a</b></i></sup></a> <a href="#cite_ref-disc1_9-1"><sup><i><b>b</b></i></sup></a></span> <span class="reference-text"><span class="citation report">Robbie Berg&#59;&#32;Lixion Avila&#32;(2012-10-22).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.001.shtml?">&#xfeff;Tropical Depression Eighteen Discussion Number 1&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.001.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.001.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-23</span>.</span></span>
</li>
<li id="cite_note-10"><span class="mw-cite-backlink"><b><a href="#cite_ref-10">^</a></b></span> <span class="reference-text"><span class="citation report">Richard Pasch&#32;(2012-10-22).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.002.shtml?">&#xfeff;Tropical Storm Sandy Discussion Number 2&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.002.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.002.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-23</span>.</span></span>
</li>
<li id="cite_note-11"><span class="mw-cite-backlink"><b><a href="#cite_ref-11">^</a></b></span> <span class="reference-text"><span class="citation report">Stacy Stewart&#32;(2012-10-23).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.003.shtml?">&#xfeff;Tropical Storm Sandy Discussion Number 3&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.003.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.003.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-23</span>.</span></span>
</li>
<li id="cite_note-12"><span class="mw-cite-backlink"><b><a href="#cite_ref-12">^</a></b></span> <span class="reference-text"><span class="citation report">Daniel Brown&#32;(2012-10-23).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.004.shtml?">&#xfeff;Tropical Storm Sandy Discussion Number 4&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.004.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.004.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-24</span>.</span></span>
</li>
<li id="cite_note-13"><span class="mw-cite-backlink"><b><a href="#cite_ref-13">^</a></b></span> <span class="reference-text"><span class="citation report">Jack Beven&#32;(2012-10-24).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.007.shtml?">&#xfeff;Tropical Storm Sandy Discussion Number 7&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.007.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.007.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-24</span>.</span></span>
</li>
<li id="cite_note-14"><span class="mw-cite-backlink"><b><a href="#cite_ref-14">^</a></b></span> <span class="reference-text"><span class="citation report">Michael Brennan&#32;(2012-10-24).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.009.shtml?">&#xfeff;Hurricane Sandy Discussion Number 9&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.009.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.009.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-24</span>.</span></span>
</li>
<li id="cite_note-15"><span class="mw-cite-backlink"><b><a href="#cite_ref-15">^</a></b></span> <span class="reference-text"><span class="citation report">Todd Kimberlain&#59;&#32;James Franklin&#32;(2002-10-24).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.update.10241918.shtml?">&#xfeff;Hurricane Sandy Tropical Cyclone Update&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.update.10241918.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.update.10241918.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-24</span>.</span></span>
</li>
<li id="cite_note-Oct_25_0430_Update-16"><span class="mw-cite-backlink">^ <a href="#cite_ref-Oct_25_0430_Update_16-0"><sup><i><b>a</b></i></sup></a> <a href="#cite_ref-Oct_25_0430_Update_16-1"><sup><i><b>b</b></i></sup></a></span> <span class="reference-text"><span class="citation web">Stewart, Stacy (October 25, 2012). <a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.update.10250441.shtml">"October 25 12:30 a.m. EDT (0430 UTC) Update"</a>. National Hurricane Center<span class="reference-accessdate">. Retrieved October 25, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.aufirst=Stacy&amp;rft.aulast=Stewart&amp;rft.au=Stewart%2C+Stacy&amp;rft.btitle=October+25+12%3A30+a.m.+EDT+%280430+UTC%29+Update&amp;rft.date=October+25%2C+2012&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.nhc.noaa.gov%2Farchive%2F2012%2Fal18%2Fal182012.update.10250441.shtml&amp;rft.pub=National+Hurricane+Center&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-17"><span class="mw-cite-backlink"><b><a href="#cite_ref-17">^</a></b></span> <span class="reference-text"><span class="citation report">Stacy Stewart&#59;&#32;Dave Roberts&#32;(October 25, 2012).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.update.10250539.shtml?">&#xfeff;Hurricane Sandy Tropical Cyclone Update&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.update.10250539.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.update.10250539.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved October 26, 2012</span>.</span></span>
</li>
<li id="cite_note-18"><span class="mw-cite-backlink"><b><a href="#cite_ref-18">^</a></b></span> <span class="reference-text"><span class="citation report">Stacy Stewart&#32;(2012-10-25).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.012.shtml?">&#xfeff;Hurricane Sandy Discussion Number 12&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.012.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.012.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-26</span>.</span></span>
</li>
<li id="cite_note-19"><span class="mw-cite-backlink"><b><a href="#cite_ref-19">^</a></b></span> <span class="reference-text"><span class="citation report">Michael Brennan&#32;(2012-10-25).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.013.shtml?">&#xfeff;Hurricane Sandy Discussion Number 13&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.013.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.013.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-26</span>.</span></span>
</li>
<li id="cite_note-20"><span class="mw-cite-backlink"><b><a href="#cite_ref-20">^</a></b></span> <span class="reference-text"><span class="citation report">Michael Brennan&#32;(2012-10-25).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.014.shtml?">&#xfeff;Hurricane Sandy Discussion Number 14&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.014.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.014.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-26</span>.</span></span>
</li>
<li id="cite_note-21"><span class="mw-cite-backlink"><b><a href="#cite_ref-21">^</a></b></span> <span class="reference-text"><span class="citation report">Jack Beven&#32;(2012-10-26).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.015.shtml?">&#xfeff;Hurricane Sandy Discussion Number 15&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.015.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.015.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-26</span>.</span></span>
</li>
<li id="cite_note-22"><span class="mw-cite-backlink"><b><a href="#cite_ref-22">^</a></b></span> <span class="reference-text"><span class="citation report">Michael Brennan&#32;(2012-10-26).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.016.shtml?">&#xfeff;Hurricane Sandy Discussion Number 16&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.016.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.016.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-26</span>.</span></span>
</li>
<li id="cite_note-23"><span class="mw-cite-backlink"><b><a href="#cite_ref-23">^</a></b></span> <span class="reference-text"><span class="citation report">Jack Beven&#32;(2012-10-27).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.019.shtml?">&#xfeff;Hurricane Sandy Discussion Number 19&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.019.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.019.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-27</span>.</span></span>
</li>
<li id="cite_note-24"><span class="mw-cite-backlink"><b><a href="#cite_ref-24">^</a></b></span> <span class="reference-text"><span class="citation report">Jack Beven&#32;(2012-10-27).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.020.shtml?">&#xfeff;Tropical Storm Sandy Discussion Number 20&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.020.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.020.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-27</span>.</span></span>
</li>
<li id="cite_note-25"><span class="mw-cite-backlink"><b><a href="#cite_ref-25">^</a></b></span> <span class="reference-text"><span class="citation report">Daniel Brown&#32;(2012-10-27).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.021.shtml?">&#xfeff;Hurricane Sandy Discussion Number 21&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.021.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.discus.021.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-27</span>.</span></span>
</li>
<li id="cite_note-26"><span class="mw-cite-backlink"><b><a href="#cite_ref-26">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.washingtonpost.com/blogs/capital-weather-gang/post/cause-for-concern-the-7-most-alarming-hurricane-sandy-images/2012/10/28/615bbbfe-210b-11e2-ac85-e669876c6a24_blog.html">"Cause for concern: the 7 most alarming Hurricane Sandy images"</a>. Washington Post<span class="reference-accessdate">. Retrieved 28 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Cause+for+concern%3A+the+7+most+alarming+Hurricane+Sandy+images&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.washingtonpost.com%2Fblogs%2Fcapital-weather-gang%2Fpost%2Fcause-for-concern-the-7-most-alarming-hurricane-sandy-images%2F2012%2F10%2F28%2F615bbbfe-210b-11e2-ac85-e669876c6a24_blog.html&amp;rft.pub=Washington+Post&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-27"><span class="mw-cite-backlink"><b><a href="#cite_ref-27">^</a></b></span> <span class="reference-text"><span class="citation report">Robbie Berg; Lixion Avila&#32;(2012-10-22).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.001.shtml?">&#xfeff;Tropical Depression Eighteen Advisory Number 1&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.001.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.001.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-24</span>.</span></span>
</li>
<li id="cite_note-28"><span class="mw-cite-backlink"><b><a href="#cite_ref-28">^</a></b></span> <span class="reference-text"><span class="citation report">Daniel Brown&#32;(2012-10-23).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.004.shtml?">&#xfeff;Tropical Storm Sandy Advisory Number 4&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.004.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.004.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-24</span>.</span></span>
</li>
<li id="cite_note-advis5-29"><span class="mw-cite-backlink">^ <a href="#cite_ref-advis5_29-0"><sup><i><b>a</b></i></sup></a> <a href="#cite_ref-advis5_29-1"><sup><i><b>b</b></i></sup></a> <a href="#cite_ref-advis5_29-2"><sup><i><b>c</b></i></sup></a></span> <span class="reference-text"><span class="citation report">Richard Pasch&#32;(2012-10-23).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.004.shtml?">&#xfeff;Tropical Storm Sandy Advisory Number 5&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.004.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.004.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-24</span>.</span></span>
</li>
<li id="cite_note-30"><span class="mw-cite-backlink"><b><a href="#cite_ref-30">^</a></b></span> <span class="reference-text"><span class="citation report">Richard Pasch&#32;(2012-10-22).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.002.shtml?">&#xfeff;Tropical Storm Sandy Advisory Number 2&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.002.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.002.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-24</span>.</span></span>
</li>
<li id="cite_note-31"><span class="mw-cite-backlink"><b><a href="#cite_ref-31">^</a></b></span> <span class="reference-text"><span class="citation report">Richard Pasch&#32;(2012-10-23).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public_a.005.shtml?">&#xfeff;Tropical Storm Sandy Advisory Number 5A&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public_a.005.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.public_a.005.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-24</span>.</span></span>
</li>
<li id="cite_note-advis6-32"><span class="mw-cite-backlink">^ <a href="#cite_ref-advis6_32-0"><sup><i><b>a</b></i></sup></a> <a href="#cite_ref-advis6_32-1"><sup><i><b>b</b></i></sup></a></span> <span class="reference-text"><span class="citation report">Richard Pasch&#32;(2012-10-23).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.006.shtml?">&#xfeff;Tropical Storm Sandy Advisory Number 6&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.006.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.006.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-24</span>.</span></span>
</li>
<li id="cite_note-33"><span class="mw-cite-backlink"><b><a href="#cite_ref-33">^</a></b></span> <span class="reference-text"><span class="citation report">John Beven II&#32;(2012-10-24).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.007.shtml?">&#xfeff;Tropical Storm Sandy Advisory Number 7&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.007.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.007.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-24</span>.</span></span>
</li>
<li id="cite_note-34"><span class="mw-cite-backlink"><b><a href="#cite_ref-34">^</a></b></span> <span class="reference-text"><span class="citation report">Richard Pasch&#32;(2012-10-26).&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.update.10261513.shtml?">&#xfeff;Hurricane Sandy Tropical Cyclone Update&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.update.10261513.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.update.10261513.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved 2012-10-26</span>.</span></span>
</li>
<li id="cite_note-35"><span class="mw-cite-backlink"><b><a href="#cite_ref-35">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.cbs12.com/news/top-stories/stories/vid_2779.shtml">"UPDATED: Hurricane Sandy closings and cancellations"</a>. cbs12.com<span class="reference-accessdate">. Retrieved 25 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=UPDATED%3A+Hurricane+Sandy+closings+and+cancellations&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.cbs12.com%2Fnews%2Ftop-stories%2Fstories%2Fvid_2779.shtml&amp;rft.pub=cbs12.com&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-36"><span class="mw-cite-backlink"><b><a href="#cite_ref-36">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.fema.gov/blog/2012-10-25/closely-monitoring-hurricane-sandy">"Closely Monitoring Hurricane Sandy"</a>. FEMA<span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Closely+Monitoring+Hurricane+Sandy&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.fema.gov%2Fblog%2F2012-10-25%2Fclosely-monitoring-hurricane-sandy&amp;rft.pub=FEMA&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-37"><span class="mw-cite-backlink"><b><a href="#cite_ref-37">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.amtrak.com/servlet/ContentServer?c=AM_Alert_C&amp;pagename=am/AM_Alert_C/Alerts_Popup&amp;cid=1251622929950">"Amtrak begins to cancel some service in advance of Hurricane Sandy"</a>. Amtrak<span class="reference-accessdate">. Retrieved 28 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Amtrak+begins+to+cancel+some+service+in+advance+of+Hurricane+Sandy&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.amtrak.com%2Fservlet%2FContentServer%3Fc%3DAM_Alert_C%26pagename%3Dam%2FAM_Alert_C%2FAlerts_Popup%26cid%3D1251622929950&amp;rft.pub=Amtrak&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-38"><span class="mw-cite-backlink"><b><a href="#cite_ref-38">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://articles.marketwatch.com/2012-10-26/finance/34742723_1_change-fees-flight-cancellations-travel-alert">"Hurricane Sandy wreaks havoc on airline flights"</a>. The Wall St. Journal<span class="reference-accessdate">. Retrieved 28 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Hurricane+Sandy+wreaks+havoc+on+airline+flights&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Farticles.marketwatch.com%2F2012-10-26%2Ffinance%2F34742723_1_change-fees-flight-cancellations-travel-alert&amp;rft.pub=The+Wall+St.+Journal&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-39"><span class="mw-cite-backlink"><b><a href="#cite_ref-39">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.bloomberg.com/news/2012-10-26/widespread-power-outages-likely-as-sandy-slams-into-east.html">"Hurricane Sandy Barrels Northward, May Hit New Jersey"</a>. Bloomberg<span class="reference-accessdate">. Retrieved 28 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Hurricane+Sandy+Barrels+Northward%2C+May+Hit+New+Jersey&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.bloomberg.com%2Fnews%2F2012-10-26%2Fwidespread-power-outages-likely-as-sandy-slams-into-east.html&amp;rft.pub=Bloomberg&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-40"><span class="mw-cite-backlink"><b><a href="#cite_ref-40">^</a></b></span> <span class="reference-text"><span class="citation web">"Trick, no treat: Forecasters predict pre-Halloween freak hurricane-winter storm mix for East". Newser.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Trick%2C+no+treat%3A+Forecasters+predict+pre-Halloween+freak+hurricane-winter+storm+mix+for+East&amp;rft.genre=book&amp;rft.pub=Newser&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span> <span style="display:none;font-size:100%" class="error citation-comment">Missing or empty <code>&#124;url=</code> (<a href="/wiki/Help:CS1_errors#cite_web_url" title="Help:CS1 errors">help</a>); </span><span style="display:none;font-size:100%" class="error citation-comment"><code>&#124;accessdate=</code> requires <code>&#124;url=</code> (<a href="/wiki/Help:CS1_errors#accessdate_missing_url" title="Help:CS1 errors">help</a>)</span></span>
</li>
<li id="cite_note-41"><span class="mw-cite-backlink"><b><a href="#cite_ref-41">^</a></b></span> <span class="reference-text"><span class="citation news">BORENSTEIN, SETH. <a rel="nofollow" class="external text" href="http://bigstory.ap.org/article/east-coast-readies-frankenstorm-monster">"EAST COAST READIES FOR 'FRANKENSTORM' MONSTER"</a>. <i>Associated Press</i><span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.atitle=EAST+COAST+READIES+FOR+%27FRANKENSTORM%27+MONSTER&amp;rft.au=BORENSTEIN%2C+SETH&amp;rft.aufirst=SETH&amp;rft.aulast=BORENSTEIN&amp;rft.genre=article&amp;rft_id=http%3A%2F%2Fbigstory.ap.org%2Farticle%2Feast-coast-readies-frankenstorm-monster&amp;rft.jtitle=Associated+Press&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-42"><span class="mw-cite-backlink"><b><a href="#cite_ref-42">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://washingtonexaminer.com/eastern-utilities-brace-for-expected-super-storm/article/feed/2042415#.UIofOm_WIm8">"Eastern utilities brace for expected super storm"</a>. The Washington Examiner<span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Eastern+utilities+brace+for+expected+super+storm&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwashingtonexaminer.com%2Feastern-utilities-brace-for-expected-super-storm%2Farticle%2Ffeed%2F2042415%23.UIofOm_WIm8&amp;rft.pub=The+Washington+Examiner&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-43"><span class="mw-cite-backlink"><b><a href="#cite_ref-43">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.wral.com/sandy-prompts-tropical-storm-watches-along-nc-coast/11703675/">"Perdue declares state of emergency before Sandy arrives"</a>. WRAL.com<span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Perdue+declares+state+of+emergency+before+Sandy+arrives&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.wral.com%2Fsandy-prompts-tropical-storm-watches-along-nc-coast%2F11703675%2F&amp;rft.pub=WRAL.com&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-autogenerated1-44"><span class="mw-cite-backlink">^ <a href="#cite_ref-autogenerated1_44-0"><sup><i><b>a</b></i></sup></a> <a href="#cite_ref-autogenerated1_44-1"><sup><i><b>b</b></i></sup></a> <a href="#cite_ref-autogenerated1_44-2"><sup><i><b>c</b></i></sup></a></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://edition.cnn.com/2012/10/26/us/tropical-weather-sandy/index.html?hpt=hp_t1">"Northeast in crosshairs of 'superstorm' Sandy"</a>. CNN.com<span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Northeast+in+crosshairs+of+%27superstorm%27+Sandy&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fedition.cnn.com%2F2012%2F10%2F26%2Fus%2Ftropical-weather-sandy%2Findex.html%3Fhpt%3Dhp_t1&amp;rft.pub=CNN.com&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-45"><span class="mw-cite-backlink"><b><a href="#cite_ref-45">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.washingtonpost.com/local/dc-mayor-declares-state-of-emergency-in-anticipation-of-hurricane-sandy/2012/10/26/88836b40-1fb3-11e2-8817-41b9a7aaa">"DC mayor declares state of emergency in anticipation of Hurricane Sandy"</a>. The Washington Post<span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=DC+mayor+declares+state+of+emergency+in+anticipation+of+Hurricane+Sandy&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.washingtonpost.com%2Flocal%2Fdc-mayor-declares-state-of-emergency-in-anticipation-of-hurricane-sandy%2F2012%2F10%2F26%2F88836b40-1fb3-11e2-8817-41b9a7aaa&amp;rft.pub=The+Washington+Post&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-46"><span class="mw-cite-backlink"><b><a href="#cite_ref-46">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.wboc.com/story/19931873/smith-island-evacuations-under-way">"Smith Island Evacuations Under Way"</a>. WBOC<span class="reference-accessdate">. Retrieved 27 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Smith+Island+Evacuations+Under+Way&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.wboc.com%2Fstory%2F19931873%2Fsmith-island-evacuations-under-way&amp;rft.pub=WBOC&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-47"><span class="mw-cite-backlink"><b><a href="#cite_ref-47">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.wboc.com/story/19932131/dorchester-county-offering-shelter-during-hurricane-sandy">"Dorchester County Offering Shelter During Hurricane Sandy"</a>. WBOC<span class="reference-accessdate">. Retrieved 27 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Dorchester+County+Offering+Shelter+During+Hurricane+Sandy&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.wboc.com%2Fstory%2F19932131%2Fdorchester-county-offering-shelter-during-hurricane-sandy&amp;rft.pub=WBOC&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-48"><span class="mw-cite-backlink"><b><a href="#cite_ref-48">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.wboc.com/story/19932081/oc-implementing-phase-i-of-emergency-operations-plan">"O.C. Implementing Phase I of Emergency Operations Plan"</a>. WBOC<span class="reference-accessdate">. Retrieved 27 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=O.C.+Implementing+Phase+I+of+Emergency+Operations+Plan&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.wboc.com%2Fstory%2F19932081%2Foc-implementing-phase-i-of-emergency-operations-plan&amp;rft.pub=WBOC&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-49"><span class="mw-cite-backlink"><b><a href="#cite_ref-49">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.wboc.com/story/19936125/obama-signs-md-emergency-declaration">"Obama Signs Md. Emergency Declaration"</a>. WBOC<span class="reference-accessdate">. Retrieved 28 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Obama+Signs+Md.+Emergency+Declaration&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.wboc.com%2Fstory%2F19936125%2Fobama-signs-md-emergency-declaration&amp;rft.pub=WBOC&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-50"><span class="mw-cite-backlink"><b><a href="#cite_ref-50">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.wmdt.com/story/19936130/worcester-county-issues-mandatory-evacuations">"Worcester County Issues Mandatory Evacuations"</a>. WMDT<span class="reference-accessdate">. Retrieved 28 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Worcester+County+Issues+Mandatory+Evacuations&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.wmdt.com%2Fstory%2F19936130%2Fworcester-county-issues-mandatory-evacuations&amp;rft.pub=WMDT&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-51"><span class="mw-cite-backlink"><b><a href="#cite_ref-51">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.wmdt.com/story/19935547/ocean-city-declares-local-state-of-emergency">"Ocean City Declares Local State Of Emergency"</a>. WMDT<span class="reference-accessdate">. Retrieved 28 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Ocean+City+Declares+Local+State+Of+Emergency&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.wmdt.com%2Fstory%2F19935547%2Focean-city-declares-local-state-of-emergency&amp;rft.pub=WMDT&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-52"><span class="mw-cite-backlink"><b><a href="#cite_ref-52">^</a></b></span> <span class="reference-text"><span class="citation web">Darryl Fears (Oct 27, 2012). <a rel="nofollow" class="external text" href="http://www.washingtonpost.com/national/health-science/sandy-poses-environmental-threat-to-chesapeake-bay/2012/10/27/64bb8630-1fb2-11e2-9cd5-b55c38388962_story.html">"Sandy poses environmental threat to Conowigo Dam"</a>. Washington Post.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.au=Darryl+Fears&amp;rft.aulast=Darryl+Fears&amp;rft.btitle=Sandy+poses+environmental+threat+to+Conowigo+Dam&amp;rft.date=Oct+27%2C+2012&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.washingtonpost.com%2Fnational%2Fhealth-science%2Fsandy-poses-environmental-threat-to-chesapeake-bay%2F2012%2F10%2F27%2F64bb8630-1fb2-11e2-9cd5-b55c38388962_story.html&amp;rft.pub=Washington+Post&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-Ocean_County-53"><span class="mw-cite-backlink"><b><a href="#cite_ref-Ocean_County_53-0">^</a></b></span> <span class="reference-text"><span class="citation news">Staff (October 27, 2012). <a rel="nofollow" class="external text" href="http://www.app.com/article/20121026/NJNEWS/310260086/Ocean-County-towns-issue-voluntary-evacuations">"Ocean County towns issue voluntary evacuations"</a>. Asbury Park Press<span class="reference-accessdate">. Retrieved October 27, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.aulast=Staff&amp;rft.au=Staff&amp;rft.btitle=Ocean+County+towns+issue+voluntary+evacuations&amp;rft.date=October+27%2C+2012&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.app.com%2Farticle%2F20121026%2FNJNEWS%2F310260086%2FOcean-County-towns-issue-voluntary-evacuations&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-54"><span class="mw-cite-backlink"><b><a href="#cite_ref-54">^</a></b></span> <span class="reference-text"><span class="citation news"><a rel="nofollow" class="external text" href="http://abclocal.go.com/wabc/story?section=weather&amp;id=8860237">"Shore towns issue voluntary evac orders"</a>. ABC7. October 26, 2012<span class="reference-accessdate">. Retrieved October 26, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Shore+towns+issue+voluntary+evac+orders&amp;rft.date=October+26%2C+2012&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fabclocal.go.com%2Fwabc%2Fstory%3Fsection%3Dweather%26id%3D8860237&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-Bay_Head-55"><span class="mw-cite-backlink"><b><a href="#cite_ref-Bay_Head_55-0">^</a></b></span> <span class="reference-text"><span class="citation news">Denise Di Stephan (October 26, 2012). <a rel="nofollow" class="external text" href="http://pointpleasant.patch.com/articles/mantoloking-advising-voluntary-evacuations">"Bay Head and Mantoloking Advising Voluntary Evacuations. Mantoloking police department urging residents to leave ahead of Sandy"</a>. Point Pleasant Patch<span class="reference-accessdate">. Retrieved October 27, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.au=Denise+Di+Stephan&amp;rft.aulast=Denise+Di+Stephan&amp;rft.btitle=Bay+Head+and+Mantoloking+Advising+Voluntary+Evacuations.+Mantoloking+police+department+urging+residents+to+leave+ahead+of+Sandy&amp;rft.date=October+26%2C+2012&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fpointpleasant.patch.com%2Farticles%2Fmantoloking-advising-voluntary-evacuations&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-56"><span class="mw-cite-backlink"><b><a href="#cite_ref-56">^</a></b></span> <span class="reference-text"><span class="citation news">[://newyork.cbslocal.com/2012/10/27/evacuations-begin-in-some-areas-of-new-jersey-for-hurricane-sandy/ "Christie Declares State of Emergency; Orders Evacuations In Some Parts of N.J."]. CBS2. October 27, 2012<span class="reference-accessdate">. Retrieved October 27, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Christie+Declares+State+of+Emergency%3B+Orders+Evacuations+In+Some+Parts+of+N.J.&amp;rft.date=October+27%2C+2012&amp;rft.genre=book&amp;rft_id=%3A%2F%2Fnewyork.cbslocal.com%2F2012%2F10%2F27%2Fevacuations-begin-in-some-areas-of-new-jersey-for-hurricane-sandy%2F&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-57"><span class="mw-cite-backlink"><b><a href="#cite_ref-57">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.nbcphiladelphia.com/news/local/Hurricane-Sandy-Philly-Mayor-urges-flood-prone-get-out-176007091.html">"Philly Mayor Tells People in Flood-Prone Areas to Leave"</a>. NBC 10 Philadelphia<span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Philly+Mayor+Tells+People+in+Flood-Prone+Areas+to+Leave&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.nbcphiladelphia.com%2Fnews%2Flocal%2FHurricane-Sandy-Philly-Mayor-urges-flood-prone-get-out-176007091.html&amp;rft.pub=NBC+10+Philadelphia&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-58"><span class="mw-cite-backlink"><b><a href="#cite_ref-58">^</a></b></span> <span class="reference-text"><span class="citation web">Mann, Ted. <a rel="nofollow" class="external text" href="http://online.wsj.com/article/SB10001424052970204840504578084582421401190.html">"New York City Subways to Shut Down - WSJ.com"</a>. Online.wsj.com<span class="reference-accessdate">. Retrieved 2012-10-28</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.aufirst=Ted&amp;rft.aulast=Mann&amp;rft.au=Mann%2C+Ted&amp;rft.btitle=New+York+City+Subways+to+Shut+Down+-+WSJ.com&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fonline.wsj.com%2Farticle%2FSB10001424052970204840504578084582421401190.html&amp;rft.pub=Online.wsj.com&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-59"><span class="mw-cite-backlink"><b><a href="#cite_ref-59">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://newyork.cbslocal.com/2012/10/26/east-coast-readies-for-frankenstorm-as-sandy-makes-its-way-north/">"Gov. Cuomo Declares State Of Emergency As Hurricane Sandy Makes Its Way North"</a>. CBS News New York<span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Gov.+Cuomo+Declares+State+Of+Emergency+As+Hurricane+Sandy+Makes+Its+Way+North&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fnewyork.cbslocal.com%2F2012%2F10%2F26%2Feast-coast-readies-for-frankenstorm-as-sandy-makes-its-way-north%2F&amp;rft.pub=CBS+News+New+York&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-60"><span class="mw-cite-backlink"><b><a href="#cite_ref-60">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.governor.ct.gov/malloy/cwp/view.asp?A=4010&amp;Q=513076">"Governor Malloy to Partially Activate the State Emergency Operations Center at 8am Saturday"</a>. <i>Hurricane Sandy News and Information</i>. CT.gov<span class="reference-accessdate">. Retrieved 27 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.atitle=Governor+Malloy+to+Partially+Activate+the+State+Emergency+Operations+Center+at+8am+Saturday&amp;rft.genre=article&amp;rft_id=http%3A%2F%2Fwww.governor.ct.gov%2Fmalloy%2Fcwp%2Fview.asp%3FA%3D4010%26Q%3D513076&amp;rft.jtitle=Hurricane+Sandy+News+and+Information&amp;rft.pub=CT.gov&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-61"><span class="mw-cite-backlink"><b><a href="#cite_ref-61">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.governor.ct.gov/malloy/cwp/view.asp?Q=513088&amp;A=4010">"Gov. Malloy Signs Declaration of Emergency"</a>. <i>Hurricane Sandy News and Information</i>. CT.gov<span class="reference-accessdate">. Retrieved 27 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.atitle=Gov.+Malloy+Signs+Declaration+of+Emergency&amp;rft.genre=article&amp;rft_id=http%3A%2F%2Fwww.governor.ct.gov%2Fmalloy%2Fcwp%2Fview.asp%3FQ%3D513088%26A%3D4010&amp;rft.jtitle=Hurricane+Sandy+News+and+Information&amp;rft.pub=CT.gov&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-chafee-62"><span class="mw-cite-backlink"><b><a href="#cite_ref-chafee_62-0">^</a></b></span> <span class="reference-text"><span class="citation web">Remiesiewicz, Neil. <a rel="nofollow" class="external text" href="http://www.wpri.com/dpp/news/local_news/west_bay/warwick-governor-urges-preparedness-for-storm">"Chafee: 'We'll be ready' for Sandy"</a>. TVL Broadcasting, Inc<span class="reference-accessdate">. Retrieved 28 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.aufirst=Neil&amp;rft.aulast=Remiesiewicz&amp;rft.au=Remiesiewicz%2C+Neil&amp;rft.btitle=Chafee%3A+%27We%27ll+be+ready%27+for+Sandy&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.wpri.com%2Fdpp%2Fnews%2Flocal_news%2Fwest_bay%2Fwarwick-governor-urges-preparedness-for-storm&amp;rft.pub=TVL+Broadcasting%2C+Inc.&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-nationalgridri-63"><span class="mw-cite-backlink">^ <a href="#cite_ref-nationalgridri_63-0"><sup><i><b>a</b></i></sup></a> <a href="#cite_ref-nationalgridri_63-1"><sup><i><b>b</b></i></sup></a></span> <span class="reference-text"><span class="citation web">Caligiuri, Courtney. <a rel="nofollow" class="external text" href="http://www.wpri.com/dpp/news/local_news/west_bay/utilities-prepare-for-possible-outages">"Utilities Prepare for Possible Outages"</a>. TVL Broadcasting, Inc<span class="reference-accessdate">. Retrieved 28 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.au=Caligiuri%2C+Courtney&amp;rft.aufirst=Courtney&amp;rft.aulast=Caligiuri&amp;rft.btitle=Utilities+Prepare+for+Possible+Outages&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.wpri.com%2Fdpp%2Fnews%2Flocal_news%2Fwest_bay%2Futilities-prepare-for-possible-outages&amp;rft.pub=TVL+Broadcasting%2C+Inc.&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-64"><span class="mw-cite-backlink"><b><a href="#cite_ref-64">^</a></b></span> <span class="reference-text"><span class="citation news"><a rel="nofollow" class="external text" href="http://www.cnn.com/2012/10/24/world/americas/tropical-weather-sandy/index.html">"Hurricane Sandy kills one in Jamaica"</a>. CNN<span class="reference-accessdate">. Retrieved 25 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Hurricane+Sandy+kills+one+in+Jamaica&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.cnn.com%2F2012%2F10%2F24%2Fworld%2Famericas%2Ftropical-weather-sandy%2Findex.html&amp;rft.pub=CNN&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-65"><span class="mw-cite-backlink"><b><a href="#cite_ref-65">^</a></b></span> <span class="reference-text"><span class="citation web">Edwards, Holly. <a rel="nofollow" class="external text" href="http://www.penbaypilot.com/article/gov-lepage-signs-emergency-declaration-ahead-sandy/4958">"Gov. LePage signs emergency declaration ahead of Sandy"</a>. penbaypilot.com<span class="reference-accessdate">. Retrieved 27 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.au=Edwards%2C+Holly&amp;rft.aufirst=Holly&amp;rft.aulast=Edwards&amp;rft.btitle=Gov.+LePage+signs+emergency+declaration+ahead+of+Sandy&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.penbaypilot.com%2Farticle%2Fgov-lepage-signs-emergency-declaration-ahead-sandy%2F4958&amp;rft.pub=penbaypilot.com&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-66"><span class="mw-cite-backlink"><b><a href="#cite_ref-66">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.foxnews.com/politics/2012/10/27/hurricane-sandy-scrambles-campaign-schedule-romney-cancels-virginia-rallies/">"Hurricane Sandy scrambles campaign schedule, Romney cancels Virginia rallies"</a>. Fox News<span class="reference-accessdate">. Retrieved 28 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Hurricane+Sandy+scrambles+campaign+schedule%2C+Romney+cancels+Virginia+rallies&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.foxnews.com%2Fpolitics%2F2012%2F10%2F27%2Fhurricane-sandy-scrambles-campaign-schedule-romney-cancels-virginia-rallies%2F&amp;rft.pub=Fox+News&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-67"><span class="mw-cite-backlink"><b><a href="#cite_ref-67">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.chicagotribune.com/travel/sns-rt-us-usa-campaign-obama-travelbre89r01a-20121027,0,6494814.story">"Obama cancels stops in Virginia, Colorado because of storm"</a>. Chicago Tribune<span class="reference-accessdate">. Retrieved 28 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Obama+cancels+stops+in+Virginia%2C+Colorado+because+of+storm&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.chicagotribune.com%2Ftravel%2Fsns-rt-us-usa-campaign-obama-travelbre89r01a-20121027%2C0%2C6494814.story&amp;rft.pub=Chicago+Tribune&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-68"><span class="mw-cite-backlink"><b><a href="#cite_ref-68">^</a></b></span> <span class="reference-text"><a rel="nofollow" class="external free" href="http://www.huffingtonpost.com/2012/10/28/hurricane-sandy-flight-ca_n_2034717.html">http://www.huffingtonpost.com/2012/10/28/hurricane-sandy-flight-ca_n_2034717.html</a></span>
</li>
<li id="cite_note-69"><span class="mw-cite-backlink"><b><a href="#cite_ref-69">^</a></b></span> <span class="reference-text"><a rel="nofollow" class="external free" href="http://www.gsmarena.com/google_cancels_october_29_event_due_to_hurricane-news-5007.php">http://www.gsmarena.com/google_cancels_october_29_event_due_to_hurricane-news-5007.php</a></span>
</li>
<li id="cite_note-70"><span class="mw-cite-backlink"><b><a href="#cite_ref-70">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.opm.gov/status/">"Federal Government Operating Status Washington DC Area"</a><span class="reference-accessdate">. Retrieved October 28, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Federal+Government+Operating+Status+Washington+DC+Area&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.opm.gov%2Fstatus%2F&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-71"><span class="mw-cite-backlink"><b><a href="#cite_ref-71">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.washingtonpost.com/local/metro-system-to-shut-down-on-monday/2012/10/28/2d1ee2be-2154-11e2-ac85-e669876c6a24_story.html?hpid=z2">"Metro system to shut down on Monday"</a>. <a href="/wiki/Washington_Post" title="Washington Post" class="mw-redirect">Washington Post</a><span class="reference-accessdate">. Retrieved October 28, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Metro+system+to+shut+down+on+Monday&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.washingtonpost.com%2Flocal%2Fmetro-system-to-shut-down-on-monday%2F2012%2F10%2F28%2F2d1ee2be-2154-11e2-ac85-e669876c6a24_story.html%3Fhpid%3Dz2&amp;rft.pub=Washington+Post&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-72"><span class="mw-cite-backlink"><b><a href="#cite_ref-72">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.weatheroffice.gc.ca/hurricane/statements_e.html">"Weather - Canadian Hurricane Information Statements"</a>. Weatheroffice.gc.ca. 2012-07-30<span class="reference-accessdate">. Retrieved 2012-10-27</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Weather+-+Canadian+Hurricane+Information+Statements&amp;rft.date=2012-07-30&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.weatheroffice.gc.ca%2Fhurricane%2Fstatements_e.html&amp;rft.pub=Weatheroffice.gc.ca&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-73"><span class="mw-cite-backlink"><b><a href="#cite_ref-73">^</a></b></span> <span class="reference-text"><span class="citation web">Junker, Wes (2011-10-18). <a rel="nofollow" class="external text" href="http://www.washingtonpost.com/blogs/capital-weather-gang/post/overnight-models-continue-to-advertise-blockbuster-storm-threat-from-hurricane-sandy/2012/10/25/b29d3980-1ea8-11e2-9cd5-b55c38388962_blog.html">"Overnight models still indicate blockbuster storm threat from Hurricane Sandy - Capital Weather Gang"</a>. The Washington Post<span class="reference-accessdate">. Retrieved 2012-10-27</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.aufirst=Wes&amp;rft.au=Junker%2C+Wes&amp;rft.aulast=Junker&amp;rft.btitle=Overnight+models+still+indicate+blockbuster+storm+threat+from+Hurricane+Sandy+-+Capital+Weather+Gang&amp;rft.date=2011-10-18&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.washingtonpost.com%2Fblogs%2Fcapital-weather-gang%2Fpost%2Fovernight-models-continue-to-advertise-blockbuster-storm-threat-from-hurricane-sandy%2F2012%2F10%2F25%2Fb29d3980-1ea8-11e2-9cd5-b55c38388962_blog.html&amp;rft.pub=The+Washington+Post&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-74"><span class="mw-cite-backlink"><b><a href="#cite_ref-74">^</a></b></span> <span class="reference-text"><span class="citation web">Chris Scott (October 26, 2012). <a rel="nofollow" class="external text" href="http://www.theweathernetwork.com/news/storm_watch_stories3&amp;stormfile=Sandy_shaping_up_to_be_an__historic__storm_26_10_2012?ref=activeweather_home">"Sandy shaping up to be an 'historic' storm"</a>. The Weather Network<span class="reference-accessdate">. Retrieved October 27, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.au=Chris+Scott&amp;rft.aulast=Chris+Scott&amp;rft.btitle=Sandy+shaping+up+to+be+an+%27historic%27+storm&amp;rft.date=October+26%2C+2012&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.theweathernetwork.com%2Fnews%2Fstorm_watch_stories3%26stormfile%3DSandy_shaping_up_to_be_an&#95;_historic&#95;_storm_26_10_2012%3Fref%3Dactiveweather_home&amp;rft.pub=The+Weather+Network&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-bahd-75"><span class="mw-cite-backlink">^ <a href="#cite_ref-bahd_75-0"><sup><i><b>a</b></i></sup></a> <a href="#cite_ref-bahd_75-1"><sup><i><b>b</b></i></sup></a></span> <span class="reference-text"><span class="citation news">Dana Smith (October 27, 2012). <a rel="nofollow" class="external text" href="http://www.tribune242.com/news/2012/oct/27/bahamas-back-business-after-hurricane-sandy/">"Bahamas Back In Business After Hurricane Sandy"</a>. <i>The Tribune (Nassau)</i><span class="reference-accessdate">. Retrieved October 28, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.atitle=Bahamas+Back+In+Business+After+Hurricane+Sandy&amp;rft.au=Dana+Smith&amp;rft.aulast=Dana+Smith&amp;rft.date=October+27%2C+2012&amp;rft.genre=article&amp;rft_id=http%3A%2F%2Fwww.tribune242.com%2Fnews%2F2012%2Foct%2F27%2Fbahamas-back-business-after-hurricane-sandy%2F&amp;rft.jtitle=The+Tribune+%28Nassau%29&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-Deaths-76"><span class="mw-cite-backlink">^ <a href="#cite_ref-Deaths_76-0"><sup><i><b>a</b></i></sup></a> <a href="#cite_ref-Deaths_76-1"><sup><i><b>b</b></i></sup></a> <a href="#cite_ref-Deaths_76-2"><sup><i><b>c</b></i></sup></a> <a href="#cite_ref-Deaths_76-3"><sup><i><b>d</b></i></sup></a></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.cnbc.com/id/49572712">"Hurricane Sandy kills 41 in Caribbean, targets U.S."</a>. <i>Reuters</i>. Nassau, Bahamas: CNBC. October 26, 2012<span class="reference-accessdate">. Retrieved October 26, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.atitle=Hurricane+Sandy+kills+41+in+Caribbean%2C+targets+U.S.&amp;rft.date=October+26%2C+2012&amp;rft.genre=article&amp;rft_id=http%3A%2F%2Fwww.cnbc.com%2Fid%2F49572712&amp;rft.jtitle=Reuters&amp;rft.place=Nassau%2C+Bahamas&amp;rft.pub=CNBC&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-charles1-77"><span class="mw-cite-backlink">^ <a href="#cite_ref-charles1_77-0"><sup><i><b>a</b></i></sup></a> <a href="#cite_ref-charles1_77-1"><sup><i><b>b</b></i></sup></a> <a href="#cite_ref-charles1_77-2"><sup><i><b>c</b></i></sup></a></span> <span class="reference-text"><span class="citation web">CHARLES AND MORGAN, JACQUELINE AND CURTIS. <a rel="nofollow" class="external text" href="http://www.miamiherald.com/2012/10/27/3069750/sandy-still-a-hurricane-heads.html">"Sandy's death toll climbs to 44 in Haiti"</a>. The Miami Herald<span class="reference-accessdate">. Retrieved 28 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.au=CHARLES+AND+MORGAN%2C+JACQUELINE+AND+CURTIS&amp;rft.aufirst=JACQUELINE+AND+CURTIS&amp;rft.aulast=CHARLES+AND+MORGAN&amp;rft.btitle=Sandy%27s+death+toll+climbs+to+44+in+Haiti&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.miamiherald.com%2F2012%2F10%2F27%2F3069750%2Fsandy-still-a-hurricane-heads.html&amp;rft.pub=The+Miami+Herald&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-78"><span class="mw-cite-backlink"><b><a href="#cite_ref-78">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.cbsnews.com/8301-202_162-57539172/hurricane-sandy-pounds-jamaica-haiti-eyes-u.s/">"Hurricane Sandy pounds Jamaica, eyes U.S."</a>. CBS<span class="reference-accessdate">. Retrieved October 24, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Hurricane+Sandy+pounds+Jamaica%2C+eyes+U.S.&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.cbsnews.com%2F8301-202_162-57539172%2Fhurricane-sandy-pounds-jamaica-haiti-eyes-u.s%2F&amp;rft.pub=CBS&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-autogenerated2-79"><span class="mw-cite-backlink">^ <a href="#cite_ref-autogenerated2_79-0"><sup><i><b>a</b></i></sup></a> <a href="#cite_ref-autogenerated2_79-1"><sup><i><b>b</b></i></sup></a></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.adn.com/2012/10/25/2672037/soggy-jamaica-cleaning-up-after.html">"Soggy Jamaica cleaning up after Hurricane Sandy"</a>. Anchorage Daily News<span class="reference-accessdate">. Retrieved 25 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Soggy+Jamaica+cleaning+up+after+Hurricane+Sandy&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.adn.com%2F2012%2F10%2F25%2F2672037%2Fsoggy-jamaica-cleaning-up-after.html&amp;rft.pub=Anchorage+Daily+News&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-80"><span class="mw-cite-backlink"><b><a href="#cite_ref-80">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://news.xinhuanet.com/english/world/2012-10/26/c_131931858.htm">"Sandy kills 6 in Haiti, 11 in Cuba, causes losses in Dominican Republic"</a>. English.news.cn<span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Sandy+kills+6+in+Haiti%2C+11+in+Cuba%2C+causes+losses+in+Dominican+Republic&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fnews.xinhuanet.com%2Fenglish%2Fworld%2F2012-10%2F26%2Fc_131931858.htm&amp;rft.pub=English.news.cn&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-81"><span class="mw-cite-backlink"><b><a href="#cite_ref-81">^</a></b></span> <span class="reference-text"><span class="citation news">Castillo and Almasy, Mariano and Steve. <a rel="nofollow" class="external text" href="http://www.cnn.com/2012/10/25/world/americas/tropical-weather-sandy/index.html">"Sandy kills 11 in Cuba, then batters Bahamas"</a>. CNN<span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.au=Castillo+and+Almasy%2C+Mariano+and+Steve&amp;rft.aufirst=Mariano+and+Steve&amp;rft.aulast=Castillo+and+Almasy&amp;rft.btitle=Sandy+kills+11+in+Cuba%2C+then+batters+Bahamas&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.cnn.com%2F2012%2F10%2F25%2Fworld%2Famericas%2Ftropical-weather-sandy%2Findex.html&amp;rft.pub=CNN&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-82"><span class="mw-cite-backlink"><b><a href="#cite_ref-82">^</a></b></span> <span class="reference-text"><span class="citation news"><a rel="nofollow" class="external text" href="http://bigstory.ap.org/article/tropical-storm-tony-develops-far-out-atlantic">"HURRICANE SANDY POUNDS JAMAICA, THEN AIMS AT CUBA"</a>. <i>Associated Press</i><span class="reference-accessdate">. Retrieved 25 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.atitle=HURRICANE+SANDY+POUNDS+JAMAICA%2C+THEN+AIMS+AT+CUBA&amp;rft.genre=article&amp;rft_id=http%3A%2F%2Fbigstory.ap.org%2Farticle%2Ftropical-storm-tony-develops-far-out-atlantic&amp;rft.jtitle=Associated+Press&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-83"><span class="mw-cite-backlink"><b><a href="#cite_ref-83">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://news.sky.com/story/1002376/hurricane-sandy-brings-heavy-flooding-to-jamaica">"Hurricane Sandy Brings Heavy Flooding To Jamaica"</a>. Skynews<span class="reference-accessdate">. Retrieved 25 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Hurricane+Sandy+Brings+Heavy+Flooding+To+Jamaica&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fnews.sky.com%2Fstory%2F1002376%2Fhurricane-sandy-brings-heavy-flooding-to-jamaica&amp;rft.pub=Skynews&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-84"><span class="mw-cite-backlink"><b><a href="#cite_ref-84">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.reuters.com/article/2012/10/25/us-storm-sandy-hurricane-idUSBRE89N16J20121025">"Hurricane Sandy pounds Jamaica, heads toward Cuba"</a>. Reuters<span class="reference-accessdate">. Retrieved 25 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Hurricane+Sandy+pounds+Jamaica%2C+heads+toward+Cuba&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.reuters.com%2Farticle%2F2012%2F10%2F25%2Fus-storm-sandy-hurricane-idUSBRE89N16J20121025&amp;rft.pub=Reuters&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-85"><span class="mw-cite-backlink"><b><a href="#cite_ref-85">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.chicagotribune.com/business/sns-rt-us-storm-sandy-hurricanebre89n16j-20121024,0,600093.story">"Hurricane Sandy growing stronger as it nears Cuba"</a>. The Chicago Tribune<span class="reference-accessdate">. Retrieved 25 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Hurricane+Sandy+growing+stronger+as+it+nears+Cuba&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.chicagotribune.com%2Fbusiness%2Fsns-rt-us-storm-sandy-hurricanebre89n16j-20121024%2C0%2C600093.story&amp;rft.pub=The+Chicago+Tribune&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-86"><span class="mw-cite-backlink"><b><a href="#cite_ref-86">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://news.blogs.cnn.com/2012/10/25/sandy-becomes-a-strong-category-2-hurricane/">"Sandy makes landfall, becomes a Category 2 Hurricane"</a>. CNN<span class="reference-accessdate">. Retrieved 25 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Sandy+makes+landfall%2C+becomes+a+Category+2+Hurricane&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fnews.blogs.cnn.com%2F2012%2F10%2F25%2Fsandy-becomes-a-strong-category-2-hurricane%2F&amp;rft.pub=CNN&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-Hurricane_Sandy_Update_Oct_25_0530_UTC-87"><span class="mw-cite-backlink"><b><a href="#cite_ref-Hurricane_Sandy_Update_Oct_25_0530_UTC_87-0">^</a></b></span> <span class="reference-text"><span class="citation web">Stacy Stewart and Dave Roberts (October 25, 2012). <a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.update.10250539.shtml">"Hurricane Sandy Update October 25 1:30 a.m. (0530 UTC)"</a>. National Hurricane Center<span class="reference-accessdate">. Retrieved October 25, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.aulast=Stacy+Stewart+and+Dave+Roberts&amp;rft.au=Stacy+Stewart+and+Dave+Roberts&amp;rft.btitle=Hurricane+Sandy+Update+October+25+1%3A30+a.m.+%280530+UTC%29&amp;rft.date=October+25%2C+2012&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.nhc.noaa.gov%2Farchive%2F2012%2Fal18%2Fal182012.update.10250539.shtml&amp;rft.pub=National+Hurricane+Center&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-88"><span class="mw-cite-backlink"><b><a href="#cite_ref-88">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.reuters.com/article/2012/10/25/us-storm-sandy-hurricane-idUSBRE89N16J20121025">"Strengthening Hurricane Sandy slams into Cuba"</a>. Reuters<span class="reference-accessdate">. Retrieved 25 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Strengthening+Hurricane+Sandy+slams+into+Cuba&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.reuters.com%2Farticle%2F2012%2F10%2F25%2Fus-storm-sandy-hurricane-idUSBRE89N16J20121025&amp;rft.pub=Reuters&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-89"><span class="mw-cite-backlink"><b><a href="#cite_ref-89">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.ajc.com/ap/ap/top-news/cuban-state-media-say-11-deaths-in-hurricane-sandy/nSn3m/">"Cuban state media say 11 deaths in Hurricane Sandy"</a>. <a href="/wiki/Atlanta_Journal-Constitution" title="Atlanta Journal-Constitution" class="mw-redirect">Atlanta Journal-Constitution</a><span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Cuban+state+media+say+11+deaths+in+Hurricane+Sandy&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.ajc.com%2Fap%2Fap%2Ftop-news%2Fcuban-state-media-say-11-deaths-in-hurricane-sandy%2FnSn3m%2F&amp;rft.pub=Atlanta+Journal-Constitution&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-90"><span class="mw-cite-backlink"><b><a href="#cite_ref-90">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.trust.org/alertnet/news/hurricane-sandy-leaves-trail-of-destruction-in-cuba/">"Hurricane Sandy leaves trail of destruction in Cuba"</a>. Trust.org via Reuters<span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Hurricane+Sandy+leaves+trail+of+destruction+in+Cuba&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.trust.org%2Falertnet%2Fnews%2Fhurricane-sandy-leaves-trail-of-destruction-in-cuba%2F&amp;rft.pub=Trust.org+via+Reuters&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-91"><span class="mw-cite-backlink"><b><a href="#cite_ref-91">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.washingtonpost.com/world/the_americas/hurricane-sandy-whips-at-bahamas-with-wind-and-rain-will-soon-threaten-us-east-coast/2012/10/25/e63d698a-1f08-11e2-8817-41b9a7aaabc7_story.html">"Hurricane Sandy rages through Bahamas, after killing 29 in Caribbean, en route to US coast"</a>. The Washington Post<span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Hurricane+Sandy+rages+through+Bahamas%2C+after+killing+29+in+Caribbean%2C+en+route+to+US+coast&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.washingtonpost.com%2Fworld%2Fthe_americas%2Fhurricane-sandy-whips-at-bahamas-with-wind-and-rain-will-soon-threaten-us-east-coast%2F2012%2F10%2F25%2Fe63d698a-1f08-11e2-8817-41b9a7aaabc7_story.html&amp;rft.pub=The+Washington+Post&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-92"><span class="mw-cite-backlink"><b><a href="#cite_ref-92">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.kypost.com/dpps/news/Guantanamo-base-cleans-up-after-Sandy_7957079">"Guantanamo base cleans up after Sandy"</a>. KY Post<span class="reference-accessdate">. Retrieved 26 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Guantanamo+base+cleans+up+after+Sandy&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.kypost.com%2Fdpps%2Fnews%2FGuantanamo-base-cleans-up-after-Sandy_7957079&amp;rft.pub=KY+Post&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-93"><span class="mw-cite-backlink"><b><a href="#cite_ref-93">^</a></b></span> <span class="reference-text"><span class="citation report">Richard Pasch.&#32;<a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.018.shtml?">&#xfeff;Hurricane Sandy Advisory Number 18&#xfeff;</a>&#32;(Report).&#32;National Hurricane Center<span class="printonly">.&#32;<a rel="nofollow" class="external free" href="http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.018.shtml">http://www.nhc.noaa.gov/archive/2012/al18/al182012.public.018.shtml</a>?</span><span class="reference-accessdate">.&#32;Retrieved October 26, 2012</span>.</span></span>
</li>
<li id="cite_note-94"><span class="mw-cite-backlink"><b><a href="#cite_ref-94">^</a></b></span> <span class="reference-text"><span class="citation news"><a rel="nofollow" class="external text" href="http://www.tribune242.com/news/2012/oct/27/btc-post-hurricane-update/">"Btc Post Hurricane Update"</a>. <i>The Tribune (Nassau)</i>. October 27, 2012<span class="reference-accessdate">. Retrieved October 28, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.atitle=Btc+Post+Hurricane+Update&amp;rft.date=October+27%2C+2012&amp;rft.genre=article&amp;rft_id=http%3A%2F%2Fwww.tribune242.com%2Fnews%2F2012%2Foct%2F27%2Fbtc-post-hurricane-update%2F&amp;rft.jtitle=The+Tribune+%28Nassau%29&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Ajournal" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-95"><span class="mw-cite-backlink"><b><a href="#cite_ref-95">^</a></b></span> <span class="reference-text"><span class="citation news"><a rel="nofollow" class="external text" href="http://www.wsvn.com/news/articles/tracking_tropics/21008896703574/south-florida-cleans-up-after-sandy/">"South Florida cleans up after Sandy"</a>. WSVN-TV. October 26, 2012<span class="reference-accessdate">. Retrieved October 26, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=South+Florida+cleans+up+after+Sandy&amp;rft.date=October+26%2C+2012&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.wsvn.com%2Fnews%2Farticles%2Ftracking_tropics%2F21008896703574%2Fsouth-florida-cleans-up-after-sandy%2F&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-96"><span class="mw-cite-backlink"><b><a href="#cite_ref-96">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.cfnews13.com/content/news/cfnews13/news/article.html/content/news/articles/cfn/2012/10/26/_2_boaters_rescued_n.html">"2 boaters rescued near New Smyrna Beach"</a>. Central Florida News 13<span class="reference-accessdate">. Retrieved 27 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=2+boaters+rescued+near+New+Smyrna+Beach&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.cfnews13.com%2Fcontent%2Fnews%2Fcfnews13%2Fnews%2Farticle.html%2Fcontent%2Fnews%2Farticles%2Fcfn%2F2012%2F10%2F26%2F_2_boaters_rescued_n.html&amp;rft.pub=Central+Florida+News+13&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-97"><span class="mw-cite-backlink"><b><a href="#cite_ref-97">^</a></b></span> <span class="reference-text"><span class="citation web"><a rel="nofollow" class="external text" href="http://www.cfnews13.com/content/news/cfnews13/sports/article.html/content/news/articles/bhsn/2012/10/25/hurricane_sandy_canc.html">"Hurricane Sandy cancels and postpones Brevard and Volusia County Friday night football games"</a>. CFN13<span class="reference-accessdate">. Retrieved 27 October 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Hurricane+Sandy+cancels+and+postpones+Brevard+and+Volusia+County+Friday+night+football+games&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.cfnews13.com%2Fcontent%2Fnews%2Fcfnews13%2Fsports%2Farticle.html%2Fcontent%2Fnews%2Farticles%2Fbhsn%2F2012%2F10%2F25%2Fhurricane_sandy_canc.html&amp;rft.pub=CFN13&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
<li id="cite_note-Orlando-98"><span class="mw-cite-backlink"><b><a href="#cite_ref-Orlando_98-0">^</a></b></span> <span class="reference-text"><span class="citation news"><a rel="nofollow" class="external text" href="http://www.clickorlando.com/weather/Hurricane-Sandy-brings-rain-strong-winds-to-Central-Florida/-/1636950/17141130/-/43onqhz/-/index.html">"Hurricane Sandy brings rain, strong winds to Central Florida"</a>. Click Orlando. October 26, 2012<span class="reference-accessdate">. Retrieved October 26, 2012</span>.</span><span title="ctx_ver=Z39.88-2004&amp;rfr_id=info%3Asid%2Fen.wikipedia.org%3AHurricane+Sandy&amp;rft.btitle=Hurricane+Sandy+brings+rain%2C+strong+winds+to+Central+Florida&amp;rft.date=October+26%2C+2012&amp;rft.genre=book&amp;rft_id=http%3A%2F%2Fwww.clickorlando.com%2Fweather%2FHurricane-Sandy-brings-rain-strong-winds-to-Central-Florida%2F-%2F1636950%2F17141130%2F-%2F43onqhz%2F-%2Findex.html&amp;rft_val_fmt=info%3Aofi%2Ffmt%3Akev%3Amtx%3Abook" class="Z3988"><span style="display:none;">&#160;</span></span></span>
</li>
</ol></div>
<h2><span class="mw-headline" id="External_links">External links</span><span class="mw-editsection"><span class="mw-editsection-bracket">[</span><a href="/w/index.php?title=Hurricane_Sandy&amp;action=edit&amp;section=20" title="Edit section: External links">edit</a><span class="mw-editsection-bracket">]</span></span></h2>
<ul>
<li><a rel="nofollow" class="external text" href="http://www.nhc.noaa.gov/#SANDY">Current information on Hurricane Sandy</a> from the <a href="/wiki/National_Hurricane_Center" title="National Hurricane Center">National Hurricane Center</a>
</li>
<li><a rel="nofollow" class="external text" href="http://www.nasa.gov/mission_pages/hurricanes/archives/2012/h2012_Sandy.html">Satellite imagery and data of Hurricane Sandy</a> from <a href="/wiki/NASA" title="NASA">NASA</a>
</li>
</ul>
<table cellspacing="0" class="navbox" style="border-spacing:0;"><tr><td style="padding:2px;"><table cellspacing="0" class="nowraplinks collapsible autocollapse navbox-inner" style="border-spacing:0;background:transparent;color:inherit;"><tr><th scope="col" class="navbox-title" colspan="3"><div class="noprint plainlinks hlist navbar mini"><ul><li class="nv-view"><a href="/wiki/Template:2012_Atlantic_hurricane_season_buttons" title="Template:2012 Atlantic hurricane season buttons"><span title="View this template" style=";;background:none transparent;border:none;;">v</span></a></li><li class="nv-talk"><a href="/wiki/Template_talk:2012_Atlantic_hurricane_season_buttons" title="Template talk:2012 Atlantic hurricane season buttons"><span title="Discuss this template" style=";;background:none transparent;border:none;;">t</span></a></li><li class="nv-edit"><a class="external text" href="//en.wikipedia.org/w/index.php?title=Template:2012_Atlantic_hurricane_season_buttons&amp;action=edit"><span title="Edit this template" style=";;background:none transparent;border:none;;">e</span></a></li></ul></div><div style="font-size:110%;">
Tropical cyclones of the <a href="/wiki/2012_Atlantic_hurricane_season" title="2012 Atlantic hurricane season">2012 Atlantic hurricane season</a></div></th></tr><tr style="height:2px;"><td></td></tr><tr><td class="navbox-image" rowspan="1" style="width:0%;padding:0px 2px 0px 0px;"><div>
<img alt="" src="//upload.wikimedia.org/wikipedia/commons/thumb/4/4f/2012_Atlantic_hurricane_season_summary_map.png/100px-2012_Atlantic_hurricane_season_summary_map.png" width="100" height="62" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/4/4f/2012_Atlantic_hurricane_season_summary_map.png/150px-2012_Atlantic_hurricane_season_summary_map.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/4/4f/2012_Atlantic_hurricane_season_summary_map.png/200px-2012_Atlantic_hurricane_season_summary_map.png 2x" /></div></td><td colspan="2" class="navbox-list navbox-odd hlist" style="width:100%;padding:0px;"><div style="padding:0em 0.25em;">
<table>
<tr>
<div title="&#123;&#123;&#123;strength}}} Alberto"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #00faf4;"><span><a href="/wiki/Tropical_cyclone#Intensity_classifications" title="Tropical cyclone">TS</a></span></td><td><a href="/wiki/2012_Atlantic_hurricane_season#Tropical_Storm_Alberto" title="2012 Atlantic hurricane season">Alberto</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Beryl"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #00faf4;"><span><a href="/wiki/Tropical_cyclone#Intensity_classifications" title="Tropical cyclone">TS</a></span></td><td><a href="/wiki/Tropical_Storm_Beryl_(2012)" title="Tropical Storm Beryl (2012)">Beryl</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Chris"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #ffffcc;"><span><a href="/wiki/Saffir%E2%80%93Simpson_hurricane_wind_scale" title="Saffir–Simpson hurricane wind scale">1</a></span></td><td><a href="/wiki/2012_Atlantic_hurricane_season#Hurricane_Chris" title="2012 Atlantic hurricane season">Chris</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Debby"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #00faf4;"><span><a href="/wiki/Tropical_cyclone#Intensity_classifications" title="Tropical cyclone">TS</a></span></td><td><a href="/wiki/Tropical_Storm_Debby_(2012)" title="Tropical Storm Debby (2012)">Debby</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Ernesto"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #ffe775;"><span><a href="/wiki/Saffir%E2%80%93Simpson_hurricane_wind_scale" title="Saffir–Simpson hurricane wind scale">2</a></span></td><td><a href="/wiki/Hurricane_Ernesto_(2012)" title="Hurricane Ernesto (2012)">Ernesto</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Florence"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #00faf4;"><span><a href="/wiki/Tropical_cyclone#Intensity_classifications" title="Tropical cyclone">TS</a></span></td><td><a href="/wiki/2012_Atlantic_hurricane_season#Tropical_Storm_Florence" title="2012 Atlantic hurricane season">Florence</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Gordon"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #ffe775;"><span><a href="/wiki/Saffir%E2%80%93Simpson_hurricane_wind_scale" title="Saffir–Simpson hurricane wind scale">2</a></span></td><td><a href="/wiki/2012_Atlantic_hurricane_season#Hurricane_Gordon" title="2012 Atlantic hurricane season">Gordon</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Helene"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #00faf4;"><span><a href="/wiki/Tropical_cyclone#Intensity_classifications" title="Tropical cyclone">TS</a></span></td><td><a href="/wiki/2012_Atlantic_hurricane_season#Tropical_Storm_Helene" title="2012 Atlantic hurricane season">Helene</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Isaac"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #ffffcc;"><span><a href="/wiki/Saffir%E2%80%93Simpson_hurricane_wind_scale" title="Saffir–Simpson hurricane wind scale">1</a></span></td><td><a href="/wiki/Hurricane_Isaac_(2012)" title="Hurricane Isaac (2012)">Isaac</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Joyce"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #00faf4;"><span><a href="/wiki/Tropical_cyclone#Intensity_classifications" title="Tropical cyclone">TS</a></span></td><td><a href="/wiki/2012_Atlantic_hurricane_season#Tropical_Storm_Joyce" title="2012 Atlantic hurricane season">Joyce</a></td></td></div>
</tr><tr>
<div title="&#123;&#123;&#123;strength}}} Kirk"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #ffe775;"><span><a href="/wiki/Saffir%E2%80%93Simpson_hurricane_wind_scale" title="Saffir–Simpson hurricane wind scale">2</a></span></td><td><a href="/wiki/2012_Atlantic_hurricane_season#Hurricane_Kirk" title="2012 Atlantic hurricane season">Kirk</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Leslie"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #ffffcc;"><span><a href="/wiki/Saffir%E2%80%93Simpson_hurricane_wind_scale" title="Saffir–Simpson hurricane wind scale">1</a></span></td><td><a href="/wiki/Hurricane_Leslie_(2012)" title="Hurricane Leslie (2012)">Leslie</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Michael"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #ffc140;"><span><a href="/wiki/Saffir%E2%80%93Simpson_hurricane_wind_scale" title="Saffir–Simpson hurricane wind scale">3</a></span></td><td><a href="/wiki/2012_Atlantic_hurricane_season#Hurricane_Michael" title="2012 Atlantic hurricane season">Michael</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Nadine"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #ffffcc;"><span><a href="/wiki/Saffir%E2%80%93Simpson_hurricane_wind_scale" title="Saffir–Simpson hurricane wind scale">1</a></span></td><td><a href="/wiki/Hurricane_Nadine_(2012)" title="Hurricane Nadine (2012)">Nadine</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Oscar"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #00faf4;"><span><a href="/wiki/Tropical_cyclone#Intensity_classifications" title="Tropical cyclone">TS</a></span></td><td><a href="/wiki/2012_Atlantic_hurricane_season#Tropical_Storm_Oscar" title="2012 Atlantic hurricane season">Oscar</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Patty"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #00faf4;"><span><a href="/wiki/Tropical_cyclone#Intensity_classifications" title="Tropical cyclone">TS</a></span></td><td><a href="/wiki/2012_Atlantic_hurricane_season#Tropical_Storm_Patty" title="2012 Atlantic hurricane season">Patty</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Rafael"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #ffffcc;"><span><a href="/wiki/Saffir%E2%80%93Simpson_hurricane_wind_scale" title="Saffir–Simpson hurricane wind scale">1</a></span></td><td><a href="/wiki/Hurricane_Rafael_(2012)" title="Hurricane Rafael (2012)" class="mw-redirect">Rafael</a></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Sandy"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #ffc140;"><span><a href="/wiki/Saffir%E2%80%93Simpson_hurricane_wind_scale" title="Saffir–Simpson hurricane wind scale">3</a></span></td><td><strong class="selflink">Sandy</strong></td></td></div>
<div title="&#123;&#123;&#123;strength}}} Tony"><td><td><div style="width: 3.0em; text-align: center; valign: middle; margin-right: 0.25em; margin-top: 0.25em; background-color: #00faf4;"><span><a href="/wiki/Tropical_cyclone#Intensity_classifications" title="Tropical cyclone">TS</a></span></td><td><a href="/wiki/2012_Atlantic_hurricane_season#Tropical_Storm_Tony" title="2012 Atlantic hurricane season">Tony</a></td></td></div>
</tr>
</table></div></td></tr><tr style="height:2px;"><td></td></tr><tr><td class="navbox-abovebelow" colspan="3"><div>
<b><img alt="Wikipedia book" src="//upload.wikimedia.org/wikipedia/commons/thumb/8/89/Symbol_book_class2.svg/16px-Symbol_book_class2.svg.png" width="16" height="16" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/8/89/Symbol_book_class2.svg/24px-Symbol_book_class2.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/8/89/Symbol_book_class2.svg/32px-Symbol_book_class2.svg.png 2x" /> <a href="/wiki/Book:2012_Atlantic_hurricane_season" title="Book:2012 Atlantic hurricane season">Book</a></b>&#160;<span style="font-weight:bold;">&#183;</span>&#32; <b><img alt="Category" src="//upload.wikimedia.org/wikipedia/en/thumb/4/48/Folder_Hexagonal_Icon.svg/16px-Folder_Hexagonal_Icon.svg.png" width="16" height="14" srcset="//upload.wikimedia.org/wikipedia/en/thumb/4/48/Folder_Hexagonal_Icon.svg/24px-Folder_Hexagonal_Icon.svg.png 1.5x, //upload.wikimedia.org/wikipedia/en/thumb/4/48/Folder_Hexagonal_Icon.svg/32px-Folder_Hexagonal_Icon.svg.png 2x" /> <a href="/wiki/Category:2012_Atlantic_hurricane_season" title="Category:2012 Atlantic hurricane season">Category</a></b>&#160;<span style="font-weight:bold;">&#183;</span>&#32; <b><img alt="Portal" src="//upload.wikimedia.org/wikipedia/en/thumb/f/fd/Portal-puzzle.svg/16px-Portal-puzzle.svg.png" width="16" height="14" srcset="//upload.wikimedia.org/wikipedia/en/thumb/f/fd/Portal-puzzle.svg/24px-Portal-puzzle.svg.png 1.5x, //upload.wikimedia.org/wikipedia/en/thumb/f/fd/Portal-puzzle.svg/32px-Portal-puzzle.svg.png 2x" /> <a href="/wiki/Portal:Tropical_cyclones" title="Portal:Tropical cyclones">Portal</a></b>&#160;<span style="font-weight:bold;">&#183;</span>&#32; <b><img alt="WikiProject" src="//upload.wikimedia.org/wikipedia/commons/thumb/3/37/People_icon.svg/16px-People_icon.svg.png" width="16" height="16" srcset="//upload.wikimedia.org/wikipedia/commons/thumb/3/37/People_icon.svg/24px-People_icon.svg.png 1.5x, //upload.wikimedia.org/wikipedia/commons/thumb/3/37/People_icon.svg/32px-People_icon.svg.png 2x" /> <a href="/wiki/Wikipedia:WikiProject_Tropical_cyclones" title="Wikipedia:WikiProject Tropical cyclones">WikiProject</a></b>&#160;<span style="font-weight:bold;">&#183;</span>&#32; <b><a href="/wiki/File:Commons-logo.svg" class="image"><img alt="Commons-logo.svg" src="//upload.wikimedia.org/wikipedia/en/thumb/4/4a/Commons-logo.svg/16px-Commons-logo.svg.png" width="16" height="22" srcset="//upload.wikimedia.org/wikipedia/en/thumb/4/4a/Commons-logo.svg/24px-Commons-logo.svg.png 1.5x, //upload.wikimedia.org/wikipedia/en/thumb/4/4a/Commons-logo.svg/32px-Commons-logo.svg.png 2x" /></a> <a href="//commons.wikimedia.org/wiki/Category:2012_Atlantic_hurricane_season" class="extiw" title="commons:Category:2012 Atlantic hurricane season">Commons</a></b></div></td></tr></table></td></tr></table>
""")
```
