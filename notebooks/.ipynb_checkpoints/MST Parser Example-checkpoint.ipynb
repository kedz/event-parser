```python
import EventParser.parser as etp
from IPython.display import Image
from collections import defaultdict
```

# Maximum Spanning Tree (MST) Parser Example

This example was taken from <a href='http://www.seas.upenn.edu/~strctlrn/bib/PDF/nonprojectiveHLT-EMNLP2005.pdf'><i>McDonald et. al. (2005)</i></a>. Running the parser in debug mode will generate a list of png files for debugging purposes.

```python
edge_weights = defaultdict(int)
edge_weights[etp.Edge('root', 'John')] = 9
edge_weights[etp.Edge('root', 'saw')] = 10
edge_weights[etp.Edge('root', 'Mary')] = 9

edge_weights[etp.Edge('saw', 'John')] = 30
edge_weights[etp.Edge('saw', 'Mary')] = 30

edge_weights[etp.Edge('John', 'saw')] = 20
edge_weights[etp.Edge('John', 'Mary')] = 3

edge_weights[etp.Edge('Mary', 'John')] = 11
edge_weights[etp.Edge('Mary', 'saw')] = 0

edges = [e for e in edge_weights]

graph = etp.Graph(edges=edges)
parser = etp.MSTParser(edge_weights)

parser.parse(graph, debug=True)
 
images = parser._dbg_imgs    
```

```python
edge_weights = defaultdict(int)
edge_weights[etp.Edge('pagerank', 'ranking algorithm')] = 8
edge_weights[etp.Edge('map', 'table')] = 10
edge_weights[etp.Edge('pagerank', 'map')] = 6

edge_weights[etp.Edge('map', 'rank')] = 3
edge_weights[etp.Edge('rank', 'pagerank')] = 4

edge_weights[etp.Edge('rank', 'list')] = 11

edges = [e for e in edge_weights]

graph = etp.Graph(edges=edges)
parser = etp.MSTParser(edge_weights)

parser.parse(graph, debug=True)
 
images = parser._dbg_imgs    
```
